```python
import pandas as pd

# Import songs dataframe
df_songs = pd.read_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_songs_labels_tags.csv')
print(f"Loaded df_songs: {len(df_songs):,} rows, {len(df_songs.columns)} columns")

# Import albums dataframe
df_albums = pd.read_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_albums_labels_tags.csv')
print(f"Loaded df_albums: {len(df_albums):,} rows, {len(df_albums.columns)} columns")

print("\n" + "="*60)
print("df_songs columns:")
print(df_songs.columns.tolist())

print("\n" + "="*60)
print("df_albums columns:")
print(df_albums.columns.tolist())

print("\n" + "="*60)
print("Sample data:")
print("\ndf_songs:")
print(df_songs.head(3))

print("\ndf_albums:")
print(df_albums.head(3))
```

```python
import pandas as pd
import numpy as np

print("Creating new columns for df_artists...")
print("=" * 80)

# ============================================================================
# 1. AGGREGATE_LABELS - All labels from songs and albums
# ============================================================================
print("\n1. Creating 'aggregate_labels' column...")

# Get labels from songs
song_labels = df_songs.groupby('performer')['label'].apply(
    lambda x: ', '.join([label for label in x.dropna() if label])
).to_dict()

# Get labels from albums
album_labels = df_albums.groupby('performer')['label'].apply(
    lambda x: ', '.join([label for label in x.dropna() if label])
).to_dict()

def get_aggregate_labels(artist):
    labels = []
    
    # Add song labels
    if artist in song_labels:
        labels.extend([l.strip() for l in str(song_labels[artist]).split(',') if l.strip()])
    
    # Add album labels
    if artist in album_labels:
        labels.extend([l.strip() for l in str(album_labels[artist]).split(',') if l.strip()])
    
    # Remove duplicates, keep order
    unique_labels = []
    for label in labels:
        if label not in unique_labels and label != 'nan':
            unique_labels.append(label)
    
    return ', '.join(unique_labels) if unique_labels else None

df_artists['aggregate_labels'] = df_artists['performer'].apply(get_aggregate_labels)
print(f"   ✓ Complete - {df_artists['aggregate_labels'].notna().sum()} artists have labels")

# ============================================================================
# 2. FIRST_TOP_10_HIT_SONG_GENRE_TAGS
# ============================================================================
print("\n2. Creating 'first_top_10_hit_song_genre_tags' column...")

def get_first_top10_song_genres(row):
    artist = row['performer']
    first_year = row['first_year_top_10_songs']
    
    if pd.isna(first_year):
        return None
    
    # Get all top 10 songs in that first year
    first_top10_songs = df_songs[
        (df_songs['performer'] == artist) & 
        (df_songs['peak_pos'] <= 10) & 
        (df_songs['first_charting_year'] == first_year)
    ]
    
    # Collect all genre tags
    all_genres = []
    for tags in first_top10_songs['genre_tags_musicbrainz'].dropna():
        all_genres.extend([g.strip() for g in str(tags).split(',') if g.strip()])
    
    # Remove duplicates
    unique_genres = list(dict.fromkeys(all_genres))  # Preserves order
    return ', '.join(unique_genres) if unique_genres else None

df_artists['first_top_10_hit_song_genre_tags'] = df_artists.apply(get_first_top10_song_genres, axis=1)
print(f"   ✓ Complete - {df_artists['first_top_10_hit_song_genre_tags'].notna().sum()} artists have song genres")

# ============================================================================
# 3. FIRST_TOP_10_HIT_ALBUM_GENRE_TAGS
# ============================================================================
print("\n3. Creating 'first_top_10_hit_album_genre_tags' column...")

def get_first_top10_album_genres(row):
    artist = row['performer']
    first_year = row['first_year_top_10_albums']
    
    if pd.isna(first_year):
        return None
    
    # Get all top 10 albums in that first year
    first_top10_albums = df_albums[
        (df_albums['performer'] == artist) & 
        (df_albums['peak_pos'] <= 10) & 
        (df_albums['first_charting_year'] == first_year)
    ]
    
    # Collect all genre tags
    all_genres = []
    for tags in first_top10_albums['genre_tags_musicbrainz'].dropna():
        all_genres.extend([g.strip() for g in str(tags).split(',') if g.strip()])
    
    # Remove duplicates
    unique_genres = list(dict.fromkeys(all_genres))
    return ', '.join(unique_genres) if unique_genres else None

df_artists['first_top_10_hit_album_genre_tags'] = df_artists.apply(get_first_top10_album_genres, axis=1)
print(f"   ✓ Complete - {df_artists['first_top_10_hit_album_genre_tags'].notna().sum()} artists have album genres")

# ============================================================================
# 4. GENRE_TAGS_THROUGH_FIRST_TOP_10_HIT
# ============================================================================
print("\n4. Creating 'genre_tags_through_first_top_10_hit' column...")

def get_genres_through_first_top10(row):
    artist = row['performer']
    
    # Find earliest top 10 year (song OR album)
    first_top10_year = None
    years = [row['first_year_top_10_songs'], row['first_year_top_10_albums']]
    valid_years = [y for y in years if pd.notna(y)]
    
    if not valid_years:
        return None
    
    first_top10_year = int(min(valid_years))
    
    # Get ALL songs/albums up to and including that year
    songs_through_year = df_songs[
        (df_songs['performer'] == artist) & 
        (df_songs['first_charting_year'] <= first_top10_year)
    ]
    
    albums_through_year = df_albums[
        (df_albums['performer'] == artist) & 
        (df_albums['first_charting_year'] <= first_top10_year)
    ]
    
    # Collect all genre tags from both
    all_genres = []
    
    for tags in songs_through_year['genre_tags_musicbrainz'].dropna():
        all_genres.extend([g.strip() for g in str(tags).split(',') if g.strip()])
    
    for tags in albums_through_year['genre_tags_musicbrainz'].dropna():
        all_genres.extend([g.strip() for g in str(tags).split(',') if g.strip()])
    
    # Remove duplicates
    unique_genres = list(dict.fromkeys(all_genres))
    return ', '.join(unique_genres) if unique_genres else None

df_artists['genre_tags_through_first_top_10_hit'] = df_artists.apply(get_genres_through_first_top10, axis=1)
print(f"   ✓ Complete - {df_artists['genre_tags_through_first_top_10_hit'].notna().sum()} artists have combined genres")

# ============================================================================
# SAVE
# ============================================================================
print("\n5. Saving updated df_artists...")
df_artists.to_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_artists_with_tags.csv', index=False)
print("   ✓ Saved to df_artists_with_tags.csv")

# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY")
print("=" * 80)
print(f"aggregate_labels: {df_artists['aggregate_labels'].notna().sum()} artists")
print(f"first_top_10_hit_song_genre_tags: {df_artists['first_top_10_hit_song_genre_tags'].notna().sum()} artists")
print(f"first_top_10_hit_album_genre_tags: {df_artists['first_top_10_hit_album_genre_tags'].notna().sum()} artists")
print(f"genre_tags_through_first_top_10_hit: {df_artists['genre_tags_through_first_top_10_hit'].notna().sum()} artists")

print("\nSample:")
print(df_artists[['performer', 'aggregate_labels', 'first_top_10_hit_song_genre_tags', 
                   'first_top_10_hit_album_genre_tags', 'genre_tags_through_first_top_10_hit']].head(10))
```

```python
import pandas as pd
import numpy as np

print("Creating 10 new columns for df_artists...")
print("=" * 80)

# ============================================================================
# 1. TOP 10 HIT SONG IN FIRST YEAR ON CHARTS
# ============================================================================
print("\n1. Creating 'top_10_hit_song_in_first_year_on_charts'...")
df_artists['top_10_hit_song_in_first_year_on_charts'] = (
    df_artists['first_year_on_chart_songs'] == df_artists['first_year_top_10_songs']
)
count = df_artists['top_10_hit_song_in_first_year_on_charts'].sum()
print(f"   ✓ {count} artists had a top 10 song in their first year on charts")

# ============================================================================
# 2. TOP 10 HIT ALBUM IN FIRST YEAR ON CHARTS
# ============================================================================
print("\n2. Creating 'top_10_hit_album_in_first_year_on_charts'...")
df_artists['top_10_hit_album_in_first_year_on_charts'] = (
    df_artists['first_year_on_chart_albums'] == df_artists['first_year_top_10_albums']
)
count = df_artists['top_10_hit_album_in_first_year_on_charts'].sum()
print(f"   ✓ {count} artists had a top 10 album in their first year on charts")

# ============================================================================
# 3. TOP 10 HIT (SONG OR ALBUM) IN FIRST YEAR ON CHARTS
# ============================================================================
print("\n3. Creating 'top_10_hit_album_or_song_in_first_year_on_charts'...")

def had_top10_in_first_year(row):
    # Find absolute first year on charts (songs or albums)
    first_years = [row['first_song_year'], row['first_album_year']]
    valid_first_years = [y for y in first_years if pd.notna(y)]
    
    if not valid_first_years:
        return False
    
    first_year_overall = int(min(valid_first_years))
    
    # Check if they had a top 10 hit (song OR album) in that year
    had_top10_song = row['first_year_top_10_songs'] == first_year_overall
    had_top10_album = row['first_year_top_10_albums'] == first_year_overall
    
    return had_top10_song or had_top10_album

df_artists['top_10_hit_album_or_song_in_first_year_on_charts'] = df_artists.apply(had_top10_in_first_year, axis=1)
count = df_artists['top_10_hit_album_or_song_in_first_year_on_charts'].sum()
print(f"   ✓ {count} artists had a top 10 hit (song OR album) in their first year on charts")

# ============================================================================
# 4. YEARS BEFORE FIRST TOP 10 HIT SONG
# ============================================================================
print("\n4. Creating 'years_before_first_top_10_hit_song'...")

def calc_years_before_top10_song(row):
    if pd.isna(row['first_year_on_chart_songs']) or pd.isna(row['first_year_top_10_songs']):
        return None
    return int(row['first_year_top_10_songs'] - row['first_year_on_chart_songs'])

df_artists['years_before_first_top_10_hit_song'] = df_artists.apply(calc_years_before_top10_song, axis=1)
df_artists['years_before_first_top_10_hit_song'] = df_artists['years_before_first_top_10_hit_song'].astype('Int64')
print(f"   ✓ Average years before first top 10 song: {df_artists['years_before_first_top_10_hit_song'].mean():.1f}")

# ============================================================================
# 5. YEARS BEFORE FIRST TOP 10 HIT ALBUM
# ============================================================================
print("\n5. Creating 'years_before_first_top_10_hit_album'...")

def calc_years_before_top10_album(row):
    if pd.isna(row['first_year_on_chart_albums']) or pd.isna(row['first_year_top_10_albums']):
        return None
    return int(row['first_year_top_10_albums'] - row['first_year_on_chart_albums'])

df_artists['years_before_first_top_10_hit_album'] = df_artists.apply(calc_years_before_top10_album, axis=1)
df_artists['years_before_first_top_10_hit_album'] = df_artists['years_before_first_top_10_hit_album'].astype('Int64')
print(f"   ✓ Average years before first top 10 album: {df_artists['years_before_first_top_10_hit_album'].mean():.1f}")

# ============================================================================
# 6. YEARS BEFORE FIRST TOP 10 HIT (SONG OR ALBUM)
# ============================================================================
print("\n6. Creating 'years_before_first_top_10_hit_song_or_album'...")

def calc_years_before_top10_any(row):
    # Find first year on charts (any)
    first_years = [row['first_song_year'], row['first_album_year']]
    valid_first_years = [y for y in first_years if pd.notna(y)]
    
    if not valid_first_years:
        return None
    
    first_year_overall = int(min(valid_first_years))
    
    # Find first top 10 year (song OR album)
    top10_years = [row['first_year_top_10_songs'], row['first_year_top_10_albums']]
    valid_top10_years = [y for y in top10_years if pd.notna(y)]
    
    if not valid_top10_years:
        return None
    
    first_top10_year = int(min(valid_top10_years))
    
    return first_top10_year - first_year_overall

df_artists['years_before_first_top_10_hit_song_or_album'] = df_artists.apply(calc_years_before_top10_any, axis=1)
df_artists['years_before_first_top_10_hit_song_or_album'] = df_artists['years_before_first_top_10_hit_song_or_album'].astype('Int64')
print(f"   ✓ Average years before first top 10 hit: {df_artists['years_before_first_top_10_hit_song_or_album'].mean():.1f}")

# ============================================================================
# 7. TOP 10 ONE-HIT WONDER (SONG)
# ============================================================================
print("\n7. Creating 'top_10_1hit_wonder_song'...")
df_artists['top_10_1hit_wonder_song'] = (df_artists['top_10_song_count'] == 1)
count = df_artists['top_10_1hit_wonder_song'].sum()
print(f"   ✓ {count} artists are top 10 one-hit wonders (songs)")

# ============================================================================
# 8. TOP 10 ONE-HIT WONDER (ALBUM)
# ============================================================================
print("\n8. Creating 'top_10_1hit_wonder_album'...")
df_artists['top_10_1hit_wonder_album'] = (df_artists['top_10_album_count'] == 1)
count = df_artists['top_10_1hit_wonder_album'].sum()
print(f"   ✓ {count} artists are top 10 one-hit wonders (albums)")

# ============================================================================
# 9. TOP 10 HITMAKER (SONGS)
# ============================================================================
print("\n9. Creating 'top_10_hitmaker_songs'...")
df_artists['top_10_hitmaker_songs'] = (df_artists['top_10_song_count'] > 1)
count = df_artists['top_10_hitmaker_songs'].sum()
print(f"   ✓ {count} artists are top 10 hitmakers (songs)")

# ============================================================================
# 10. TOP 10 HITMAKER (ALBUMS)
# ============================================================================
print("\n10. Creating 'top_10_hitmaker_albums'...")
df_artists['top_10_hitmaker_albums'] = (df_artists['top_10_album_count'] > 1)
count = df_artists['top_10_hitmaker_albums'].sum()
print(f"   ✓ {count} artists are top 10 hitmakers (albums)")

# ============================================================================
# SAVE
# ============================================================================
print("\n11. Saving updated df_artists...")
df_artists.to_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_artists_with_tags.csv', index=False)
print("   ✓ Saved to df_artists_with_tags.csv")

# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY - New Columns Created:")
print("=" * 80)

new_cols = [
    'top_10_hit_song_in_first_year_on_charts',
    'top_10_hit_album_in_first_year_on_charts',
    'top_10_hit_album_or_song_in_first_year_on_charts',
    'years_before_first_top_10_hit_song',
    'years_before_first_top_10_hit_album',
    'years_before_first_top_10_hit_song_or_album',
    'top_10_1hit_wonder_song',
    'top_10_1hit_wonder_album',
    'top_10_hitmaker_songs',
    'top_10_hitmaker_albums'
]

for col in new_cols:
    print(f"✓ {col}")

print("\nSample:")
print(df_artists[['performer'] + new_cols].head(10))
```

```python
import pandas as pd

print("Adding one-hit wonder and hitmaker columns to songs and albums...")
print("=" * 80)

# ============================================================================
# 1. SONGS DATAFRAME
# ============================================================================
print("\n1. Processing df_songs...")

# Merge artist's top 10 song count
df_songs = df_songs.merge(
    df_artists[['performer', 'top_10_song_count']], 
    on='performer', 
    how='left',
    suffixes=('', '_artist')
)

# Create columns
df_songs['top_10_1hit_wonder_song'] = (df_songs['top_10_song_count'] == 1)
df_songs['top_10_hitmaker_song'] = (df_songs['top_10_song_count'] > 1)

# Count songs in each category
one_hit_wonder_songs = df_songs['top_10_1hit_wonder_song'].sum()
hitmaker_songs = df_songs['top_10_hitmaker_song'].sum()

print(f"   Songs by one-hit wonders: {one_hit_wonder_songs:,}")
print(f"   Songs by hitmakers: {hitmaker_songs:,}")

# Drop the merged count column (we don't need it in the final dataframe)
df_songs = df_songs.drop(columns=['top_10_song_count'], errors='ignore')

# ============================================================================
# 2. ALBUMS DATAFRAME
# ============================================================================
print("\n2. Processing df_albums...")

# Merge artist's top 10 album count
df_albums = df_albums.merge(
    df_artists[['performer', 'top_10_album_count']], 
    on='performer', 
    how='left',
    suffixes=('', '_artist')
)

# Create columns
df_albums['top_10_1hit_wonder_album'] = (df_albums['top_10_album_count'] == 1)
df_albums['top_10_hitmaker_album'] = (df_albums['top_10_album_count'] > 1)

# Count albums in each category
one_hit_wonder_albums = df_albums['top_10_1hit_wonder_album'].sum()
hitmaker_albums = df_albums['top_10_hitmaker_album'].sum()

print(f"   Albums by one-hit wonders: {one_hit_wonder_albums:,}")
print(f"   Albums by hitmakers: {hitmaker_albums:,}")

# Drop the merged count column
df_albums = df_albums.drop(columns=['top_10_album_count'], errors='ignore')

# ============================================================================
# 3. SAVE TO CSV
# ============================================================================
print("\n3. Saving to CSV...")

df_songs.to_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_songs_labels_tags.csv', index=False)
print("   ✓ Saved df_songs to df_songs_labels_tags.csv")

df_albums.to_csv('/Users/jamesemcnally/Dropbox/DSBC/Spring 2026/df_albums_labels_tags.csv', index=False)
print("   ✓ Saved df_albums to df_albums_labels_tags.csv")

# ============================================================================
# 4. SUMMARY
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY")
print("=" * 80)

print("\nSongs:")
print(f"  Total songs: {len(df_songs):,}")
print(f"  By one-hit wonders: {one_hit_wonder_songs:,} ({one_hit_wonder_songs/len(df_songs)*100:.1f}%)")
print(f"  By hitmakers: {hitmaker_songs:,} ({hitmaker_songs/len(df_songs)*100:.1f}%)")

print("\nAlbums:")
print(f"  Total albums: {len(df_albums):,}")
print(f"  By one-hit wonders: {one_hit_wonder_albums:,} ({one_hit_wonder_albums/len(df_albums)*100:.1f}%)")
print(f"  By hitmakers: {hitmaker_albums:,} ({hitmaker_albums/len(df_albums)*100:.1f}%)")

print("\nSample songs:")
print(df_songs[['performer', 'title', 'peak_pos', 'top_10_1hit_wonder_song', 'top_10_hitmaker_song']].head(10))

print("\nSample albums:")
print(df_albums[['performer', 'title', 'peak_pos', 'top_10_1hit_wonder_album', 'top_10_hitmaker_album']].head(10))

print("\n✓ Complete!")
```
