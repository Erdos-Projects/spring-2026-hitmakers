```python
import pandas as pd

df_artists = pd.read_csv('df_artists.csv')
df_songs   = pd.read_csv('df_songs.csv')
df_albums  = pd.read_csv('df_albums.csv')

print(f"df_artists: {df_artists.shape}")
print(f"df_songs:   {df_songs.shape}")
print(f"df_albums:  {df_albums.shape}")
```

```python
pd.set_option('display.max_columns', None)

print("df_artists:")
display(df_artists.head())
```

```python
# Columns to remove
cols_to_drop = [
    'last_song_year', 'last_album_year', '#_of_unique_years_active',
    'top_20_song_count', 'top_20_album_count', 'top_50_song_count', 'top_50_album_count',
    'total_charting_songs_duration_weeks', 'total_charting_albums_duration_weeks',
    'tags',
    'first_year_top_20_songs', 'first_year_top_20_albums',
    'first_top_10_song_labels', 'first_top_10_song_big3_label',
    'first_top_10_album_labels', 'first_top_10_album_big3_label',
    'top_10_hit_song_in_first_year_on_charts', 'top_10_hit_album_in_first_year_on_charts',
    'top_10_hit_album_or_song_in_first_year_on_charts',
    'years_before_first_top_10_hit_song', 'years_before_first_top_10_hit_album',
    'years_before_first_top_10_hit_song_or_album'
]

# Drop only columns that exist
existing_drops = [c for c in cols_to_drop if c in df_artists.columns]
missing_drops  = [c for c in cols_to_drop if c not in df_artists.columns]
df_artists = df_artists.drop(columns=existing_drops)
print(f"Dropped {len(existing_drops)} columns")
if missing_drops:
    print(f"Already absent (skipped): {missing_drops}")

# Convert to nullable integer (handles NaN safely)
int_cols = [
    'first_year_on_chart_songs', 'first_year_on_chart_albums',
    'first_year_top_10_songs', 'first_year_top_10_albums',
    'first_year_num1_songs', 'first_year_num1_albums'
]
for c in int_cols:
    if c in df_artists.columns:
        df_artists[c] = pd.to_numeric(df_artists[c], errors='coerce').astype('Int64')
        print(f"  Converted {c} to Int64")
    else:
        print(f"  Skipped (not found): {c}")

print(f"\ndf_artists shape: {df_artists.shape}")
```

```python
df_artists.to_csv('df_artists.csv', index=False)
print(f"Saved df_artists: {df_artists.shape}")
```
