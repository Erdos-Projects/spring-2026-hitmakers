```python
# When we originally assigned MusicBrainz integer IDs to artists in df_artists, we used
# name-based matching against the MusicBrainz database. This introduced systematic errors:
# many common artist names (e.g. "The Beatles", "Nirvana", "Madonna", "Eminem") match
# multiple MusicBrainz entries, and our lookup frequently resolved to the wrong artist —
# often an obscure band or foreign artist who happened to share the name. For example:
#   - "The Beatles" → a 1960s Philadelphia doo-wop group (id=2589354) instead of the Fab Four (id=303)
#   - "Nirvana"     → a Croatian prog-rock band (id=1706467) instead of the US grunge band (id=54)
#   - "Madonna"     → a Lebanese singer (id=1690463) instead of the Queen of Pop (id=89)
#
# We corrected this by cross-referencing against the Kang et al. dataset, which contains
# verified MusicBrainz IDs for charting artists. A spot check against the local MusicBrainz
# database dump confirmed the Kang IDs are correct in every case. We apply those corrections
# here for all 564 unambiguous conflicts (cases where the name maps to exactly one Kang entry).

update_map = conflict_candidates.set_index('name_lower')['id'].to_dict()

df_artists['name_lower'] = df_artists['name'].str.lower().str.strip()
df_artists.loc[df_artists['name_lower'].isin(update_map), 'musicbrainz_artist_id'] = (
    df_artists.loc[df_artists['name_lower'].isin(update_map), 'name_lower'].map(update_map)
)
df_artists = df_artists.drop(columns=['name_lower'])

print(f"Updated {len(update_map)} artist IDs from kang dataset.")

df_artists.to_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_artists.csv', index=False)
print("Saved.")
```

```python
import re

def normalize_name(name):
    if not isinstance(name, str):
        return ''
    n = name.lower().strip()
    # Remove punctuation except letters, numbers, spaces
    n = re.sub(r"[\"'\.&,\-\(\)]", ' ', n)
    n = re.sub(r'\s+', ' ', n).strip()
    return n

# Flag artifact names
artifact_patterns = r'^(&|feat|ft\.?|w/|x |and )'
missing['is_artifact'] = missing['name'].str.lower().str.strip().str.match(artifact_patterns)

real_missing = missing[~missing['is_artifact']].copy()
real_missing['name_norm'] = real_missing['name'].apply(normalize_name)
df_mb['name_norm'] = df_mb['name'].apply(normalize_name)

print(f"Artifacts filtered out: {missing['is_artifact'].sum()}")
print(f"Real missing artists: {len(real_missing)}")

# Try normalized join
merged_norm = real_missing[['name', 'name_norm']].merge(
    df_mb[['id', 'name_norm', 'comment']],
    on='name_norm', how='left'
)
counts = merged_norm.groupby('name_norm')['id'].count().reset_index()
counts.columns = ['name_norm', 'mb_match_count']
merged_norm = merged_norm.merge(counts, on='name_norm')

safe_norm   = merged_norm[merged_norm['mb_match_count'] == 1].drop_duplicates('name_norm')
ambig_norm  = merged_norm[merged_norm['mb_match_count'] > 1]['name_norm'].nunique()
no_match    = merged_norm[merged_norm['id'].isna()]['name_norm'].nunique()

print(f"\nUnambiguous matches: {len(safe_norm)}")
print(f"Ambiguous (multiple): {ambig_norm}")
print(f"No match:            {no_match}")
```

```python
import unicodedata

def strip_accents(s):
    if not isinstance(s, str):
        return s
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'
    ).lower().strip()

not_in_df_copy['name_norm'] = not_in_df_copy['name'].apply(strip_accents)
df_artists_names['name_norm'] = df_artists_names['name'].apply(strip_accents)

name_matches = not_in_df_copy.merge(
    df_artists_names,
    on='name_norm', how='inner'
)

print(f"Name matches (accent-insensitive): {len(name_matches)}")
print(name_matches[['name_x', 'name_y', 'id', 'musicbrainz_artist_id', 'popularity']].to_string())
```

```python
# Create combined_major_genre_categories as the union of MB and Spotify categories,
# sorted alphabetically with no duplicates.

def combine_genres(mb, sp):
    mb_set = set(c.strip() for c in mb.split(',')) if isinstance(mb, str) else set()
    sp_set = set(c.strip() for c in sp.split(',')) if isinstance(sp, str) else set()
    combined = mb_set | sp_set
    return ', '.join(sorted(combined)) if combined else None

df_artists['combined_major_genre_categories'] = df_artists.apply(
    lambda r: combine_genres(
        r['musicbrainz_major_genre_categories'],
        r['spotify_major_genre_categories']
    ), axis=1
)

filled = df_artists['combined_major_genre_categories'].notna().sum()
print(f"Artists with combined genres: {filled:,} ({filled/len(df_artists)*100:.1f}%)")

print("\nDistribution:")
all_cats = []
for cats in df_artists['combined_major_genre_categories'].dropna():
    all_cats.extend([c.strip() for c in cats.split(',')])
print(pd.Series(all_cats).value_counts().to_string())

df_artists.to_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_artists.csv', index=False)
print("\nSaved.")
```

```python
# Look up the correct UUID (musicbrainz_mbid) for each integer musicbrainz_artist_id
# using the local MusicBrainz artist dump, then sync df_artists.

import pandas as pd

artist_path = '/Users/jamesemcnally/musicbrainz_data/mbdump/artist'

# Load only id and gid columns from the MB dump (fast — just first 2 fields per line)
print("Loading MB artist id→uuid mapping...")
mb_id_to_uuid = {}
with open(artist_path, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.split('\t')
        if len(parts) >= 2:
            try:
                mb_id_to_uuid[int(parts[0])] = parts[1]
            except ValueError:
                continue

print(f"Loaded {len(mb_id_to_uuid):,} MB artists")

# Map integer IDs in df_artists to correct UUIDs
df_artists['musicbrainz_artist_id_int'] = pd.to_numeric(
    df_artists['musicbrainz_artist_id'], errors='coerce'
).astype('Int64')

df_artists['musicbrainz_mbid'] = df_artists['musicbrainz_artist_id_int'].map(
    lambda x: mb_id_to_uuid.get(x) if pd.notna(x) else None
)

df_artists = df_artists.drop(columns=['musicbrainz_artist_id_int'])

filled = df_artists['musicbrainz_mbid'].notna().sum()
print(f"musicbrainz_mbid filled: {filled:,} / {len(df_artists):,}")

df_artists.to_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_artists.csv', index=False)
print("Saved.")
```

```python
# Sync corrected MusicBrainz artist IDs from df_artists into df_songs and df_albums,
# and rename performer_normalized → name in both.
#
# df_artists uses 'name' as the key (renamed from performer_normalized earlier).
# df_songs and df_albums still use 'performer_normalized' — we match on that.

import pandas as pd

df_artists = pd.read_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_artists.csv')
df_songs   = pd.read_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_songs.csv')
df_albums  = pd.read_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_albums.csv')

# Build name → corrected ID map from df_artists
correct_id_map = df_artists.set_index('name')['musicbrainz_artist_id'].dropna().to_dict()

# Apply to df_songs
df_songs['musicbrainz_artist_id'] = df_songs['performer_normalized'].map(correct_id_map)
df_songs = df_songs.rename(columns={'performer_normalized': 'name'})

# Apply to df_albums
df_albums['musicbrainz_artist_id'] = df_albums['performer_normalized'].map(correct_id_map)
df_albums = df_albums.rename(columns={'performer_normalized': 'name'})

# Summary
print(f"df_songs  — IDs filled: {df_songs['musicbrainz_artist_id'].notna().sum()} / {len(df_songs)}")
print(f"df_albums — IDs filled: {df_albums['musicbrainz_artist_id'].notna().sum()} / {len(df_albums)}")

# Make 'name' the second column in both dataframes
def move_to_second(df):
    cols = df.columns.tolist()
    cols.insert(1, cols.pop(cols.index('name')))
    return df[cols]

df_songs  = move_to_second(df_songs)
df_albums = move_to_second(df_albums)

# Save
df_songs.to_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_songs.csv', index=False)
df_albums.to_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/df_albums.csv', index=False)
print("Saved.")
```
