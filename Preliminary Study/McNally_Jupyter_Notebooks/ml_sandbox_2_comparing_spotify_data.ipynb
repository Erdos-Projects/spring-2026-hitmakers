```python
# Load the Spotify dataset CSV into a DataFrame
import pandas as pd

katie_spotify_data = pd.read_csv('/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/spotify_dataset.csv')
katie_spotify_data.head()
```

```python
# Compare songs between katie_spotify_data and df_songs by Spotify track ID,
# and report the overlap statistics between the two datasets

katie_ids = set(katie_spotify_data['track_id'].dropna())
songs_ids = set(df_songs['spotify_id'].dropna())

in_both = katie_ids & songs_ids
only_in_katie = katie_ids - songs_ids
only_in_df_songs = songs_ids - katie_ids

print(f"katie_spotify_data unique track_ids:  {len(katie_ids):,}")
print(f"df_songs unique spotify_ids:          {len(songs_ids):,}")
print(f"")
print(f"In both (overlap):                    {len(in_both):,}")
print(f"Only in katie_spotify_data:           {len(only_in_katie):,}")
print(f"Only in df_songs:                     {len(only_in_df_songs):,}")
print(f"")
print(f"Overlap as % of katie_spotify_data:   {len(in_both)/len(katie_ids)*100:.1f}%")
print(f"Overlap as % of df_songs:             {len(in_both)/len(songs_ids)*100:.1f}%")
```

```python
# Create df_songs_top_20 containing only songs that reached the top 20 on the Billboard Hot 100
df_songs_top_20 = df_songs[df_songs['peak_pos'] <= 20].copy()

print(f"Total songs in df_songs:  {len(df_songs):,}")
print(f"Songs that hit top 20:    {len(df_songs_top_20):,}")
df_songs_top_20.head()
```

```python
# Check overlap between katie_spotify_data and df_songs_top_20:
# first by Spotify ID, then by song title + artist name for unmatched songs.
# Adjust 'track_name' and 'artists' if katie_spotify_data uses different column names.

# --- Step 1: Match by Spotify ID ---
top20_ids = set(df_songs_top_20['spotify_id'].dropna())
katie_ids = set(katie_spotify_data['track_id'].dropna())

matched_by_id = top20_ids & katie_ids

print(f"df_songs_top_20 unique spotify_ids:       {len(top20_ids):,}")
print(f"katie_spotify_data unique track_ids:      {len(katie_ids):,}")
print(f"Matched by ID:                            {len(matched_by_id):,}")
print(f"  → {len(matched_by_id)/len(top20_ids)*100:.1f}% of top-20 songs found in katie's data")
print()

# --- Step 2: Name+artist match for top-20 songs NOT matched by ID ---
unmatched_top20 = df_songs_top_20[~df_songs_top_20['spotify_id'].isin(matched_by_id)].copy()

katie_name_key = (
    katie_spotify_data['track_name'].str.strip().str.lower() + '|||' +
    katie_spotify_data['artists'].str.strip().str.lower()
)
katie_name_set = set(katie_name_key.dropna())

unmatched_top20['_match_key'] = (
    unmatched_top20['title'].str.strip().str.lower() + '|||' +
    unmatched_top20['name'].str.strip().str.lower()
)
matched_by_name = unmatched_top20['_match_key'].isin(katie_name_set).sum()

print(f"Top-20 songs not matched by ID:           {len(unmatched_top20):,}")
print(f"Of those, matched by title+artist name:   {matched_by_name:,}")
print()

total_matched = len(matched_by_id) + matched_by_name
print(f"Total top-20 songs found in katie's data: {total_matched:,} / {len(df_songs_top_20):,} ({total_matched/len(df_songs_top_20)*100:.1f}%)")
```

```python
# Check how many songs in df_songs_top_20 have non-null danceability values
total = len(df_songs_top_20)
non_null = df_songs_top_20['danceability'].notna().sum()
null = df_songs_top_20['danceability'].isna().sum()

print(f"Total songs in df_songs_top_20:    {total:,}")
print(f"Non-null danceability:             {non_null:,} ({non_null/total*100:.1f}%)")
print(f"Null danceability:                 {null:,} ({null/total*100:.1f}%)")
```

```python
# Explore non-null danceability coverage in df_songs_top_20, broken down by decade
df_songs_top_20['decade'] = (df_songs_top_20['first_charting_year'] // 10 * 10).astype('Int64')

decade_summary = (
    df_songs_top_20.groupby('decade')['danceability']
    .agg(
        total='count',
        non_null=lambda x: x.notna().sum(),
        null=lambda x: x.isna().sum()
    )
    .assign(pct_non_null=lambda df: (df['non_null'] / df['total'] * 100).round(1))
)

# 'total' from agg count only counts non-nulls, so recalculate from group size
decade_summary['total'] = df_songs_top_20.groupby('decade')['danceability'].size()
decade_summary['pct_non_null'] = (decade_summary['non_null'] / decade_summary['total'] * 100).round(1)

print(decade_summary.to_string())
```

```python
# For songs in df_songs_top_20 with null danceability, check how many have
# non-null danceability in katie_spotify_data (matched by Spotify ID)

null_dance = df_songs_top_20[df_songs_top_20['danceability'].isna()].copy()

# Merge with katie_spotify_data on spotify ID
null_merged = null_dance.merge(
    katie_spotify_data[['track_id', 'danceability']].rename(columns={'danceability': 'katie_danceability'}),
    left_on='spotify_id',
    right_on='track_id',
    how='left'
)

recoverable = null_merged['katie_danceability'].notna().sum()
still_null   = null_merged['katie_danceability'].isna().sum()

print(f"Songs in df_songs_top_20 with null danceability:   {len(null_dance):,}")
print(f"Of those, katie has a non-null value:              {recoverable:,} ({recoverable/len(null_dance)*100:.1f}%)")
print(f"Still null after checking katie's data:            {still_null:,} ({still_null/len(null_dance)*100:.1f}%)")
```
