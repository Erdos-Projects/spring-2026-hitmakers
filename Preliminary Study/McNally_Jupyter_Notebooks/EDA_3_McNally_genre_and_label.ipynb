```python
import pandas as pd

# Read the dataset
df = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

# Extract all unique genre tags
all_genres = set()
for genres in df['genre_tags_musicbrainz'].dropna():
    if isinstance(genres, str):
        genre_list = [g.strip() for g in genres.split(',')]
        all_genres.update(genre_list)

# Sort and display
unique_genres = sorted(all_genres)
print(f"Total unique genres: {len(unique_genres)}\n")
for genre in unique_genres:
    print(genre)
```
