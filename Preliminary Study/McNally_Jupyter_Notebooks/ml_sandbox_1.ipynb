```python
# Load project dataframes
import pandas as pd

base = '/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/'

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 50)

df_artists                 = pd.read_csv(base + 'df_artists.csv')
df_songs                   = pd.read_csv(base + 'df_songs.csv')
df_artists_network_metrics = pd.read_csv(base + 'df_artists_network_metrics.csv')

print(f'df_artists:                 {df_artists.shape}')
print(f'df_songs:                   {df_songs.shape}')
print(f'df_artists_network_metrics: {df_artists_network_metrics.shape}')
```

```python
# Build df_ml_sandbox
import pandas as pd
import numpy as np

# ── Base: artists with at least 1 top 10 song ─────────────────────────────────
base_cols = ['name', 'top_10_song_count', 'combined_major_genre_categories',
             'first_year_top_10_songs']
df_ml_sandbox = (df_artists[base_cols]
                 .query('top_10_song_count >= 1')
                 .copy()
                 .reset_index(drop=True))

# ── Feature 1: year of first top 10 hit ───────────────────────────────────────
df_ml_sandbox = df_ml_sandbox.rename(columns={'first_year_top_10_songs': 'first_top10_year'})

# ── Feature 2: songs_through_first_top_10_hit ─────────────────────────────────
# Count all songs that charted at any level up to and including the year
# of the artist's first top 10 hit
first_top10_year_map = df_ml_sandbox.set_index('name')['first_top10_year'].to_dict()

songs_with_year = df_songs[df_songs['name'].isin(first_top10_year_map)].copy()
songs_with_year['first_top10_year'] = songs_with_year['name'].map(first_top10_year_map)
songs_through = (songs_with_year[songs_with_year['first_charting_year'] <= songs_with_year['first_top10_year']]
                 .groupby('name')
                 .size()
                 .rename('songs_through_first_top_10_hit'))

df_ml_sandbox = df_ml_sandbox.merge(songs_through, on='name', how='left')

# ── Feature 3: #_of_major_genre_categories ────────────────────────────────────
def count_genres(s):
    if pd.isna(s) or str(s).strip() == '':
        return np.nan
    return len([x for x in str(s).split(',') if x.strip()])

df_ml_sandbox['#_of_major_genre_categories'] = (
    df_ml_sandbox['combined_major_genre_categories'].apply(count_genres))

# ── Feature 4: major_label (at time of first top 10 song) ────────────────────
first_top10_song = (df_songs[df_songs['peak_pos'] <= 10]
                    .sort_values('first_charting_year')
                    .groupby('name', as_index=False)
                    .first()
                    [['name', 'warner', 'universal', 'sony']])
first_top10_song['major_label'] = (
    first_top10_song['warner'] |
    first_top10_song['universal'] |
    first_top10_song['sony']
).astype(int)

df_ml_sandbox = df_ml_sandbox.merge(
    first_top10_song[['name', 'major_label']], on='name', how='left')

# ── Feature 5: network metrics (fillna 0) ─────────────────────────────────────
nm_cols = {
    'power_of_connected_artists_top10_rolling5':    'power_r5',
    'eigenvector_centrality_top10_rolling5':         'eigenvector_r5',
    'betweenness_centrality_top10_rolling5':         'betweenness_r5',
    'harmonic_closeness_centrality_top10_rolling5':  'harmonic_r5',
    'degree_centrality_top10_rolling5':              'degree_r5',
}
nm_sub = (df_artists_network_metrics
          [['name'] + list(nm_cols.keys())]
          .rename(columns=nm_cols))

df_ml_sandbox = df_ml_sandbox.merge(nm_sub, on='name', how='left')
for col in nm_cols.values():
    df_ml_sandbox[col] = df_ml_sandbox[col].fillna(0)

# ── Target: hitmaker ──────────────────────────────────────────────────────────
df_ml_sandbox['hitmaker'] = (df_ml_sandbox['top_10_song_count'] > 1).astype(int)
df_ml_sandbox = df_ml_sandbox.drop(columns=['top_10_song_count'])

# ── Final column order ────────────────────────────────────────────────────────
df_ml_sandbox = df_ml_sandbox[[
    'name', 'first_top10_year',
    'songs_through_first_top_10_hit',
    'combined_major_genre_categories', '#_of_major_genre_categories',
    'major_label',
    'power_r5', 'eigenvector_r5', 'betweenness_r5', 'harmonic_r5', 'degree_r5',
    'hitmaker'
]]

# ── Summary stats ─────────────────────────────────────────────────────────────
wonders = df_ml_sandbox[df_ml_sandbox['hitmaker'] == 0]
makers  = df_ml_sandbox[df_ml_sandbox['hitmaker'] == 1]

print(f"df_ml_sandbox shape: {df_ml_sandbox.shape}")
print(f"1-hit wonders: {len(wonders)} ({len(wonders)/len(df_ml_sandbox)*100:.1f}%)")
print(f"Hitmakers:     {len(makers)} ({len(makers)/len(df_ml_sandbox)*100:.1f}%)")

print(f"\n── Null rates ──")
print(df_ml_sandbox.isnull().mean().round(3).to_string())

numeric_cols = ['first_top10_year', 'songs_through_first_top_10_hit',
                '#_of_major_genre_categories', 'power_r5',
                'eigenvector_r5', 'betweenness_r5', 'harmonic_r5', 'degree_r5']

for label, grp in [('1-HIT WONDERS', wonders), ('HITMAKERS', makers)]:
    print(f"\n── {label} (n={len(grp)}) ──")
    print(grp[numeric_cols].describe().round(3).to_string())

print(f"\n── major_label ──")
print(pd.crosstab(df_ml_sandbox['major_label'],
                  df_ml_sandbox['hitmaker'].map({0:'1HW', 1:'Hitmaker'}),
                  margins=True))
```

```python
# Binary-encode combined_major_genre_categories (multi-label: "Pop, Rock" → binary columns)

# Get all unique genre categories across all artists
all_genres = sorted(set(
    genre.strip()
    for cats in df_ml_sandbox['combined_major_genre_categories'].dropna()
    for genre in cats.split(',')
    if genre.strip()
))

print(f"Unique genre categories ({len(all_genres)}):")
print(all_genres)

# Create one binary column per genre
# NaN where combined_major_genre_categories is missing (unknown, not "absent")
for genre in all_genres:
    col = 'genre_' + genre.lower().replace('/', '_').replace(' ', '_').replace(',', '').replace('-', '_')
    df_ml_sandbox[col] = df_ml_sandbox['combined_major_genre_categories'].apply(
        lambda s: int(genre in [x.strip() for x in s.split(',')]) if pd.notna(s) else np.nan
    )

genre_cols = [c for c in df_ml_sandbox.columns if c.startswith('genre_')]
print(f"\n{len(genre_cols)} genre columns added")
print(f"Genre null rate: {df_ml_sandbox[genre_cols[0]].isna().mean():.1%} (where combined_major_genre_categories is missing)")

# Value counts per genre, split by hitmaker status
print("\n── Genre prevalence by group ──")
summary = []
for col in genre_cols:
    sub = df_ml_sandbox[df_ml_sandbox[col].notna()]
    pct_1hw = 100 * sub[sub['hitmaker']==0][col].mean()
    pct_hm  = 100 * sub[sub['hitmaker']==1][col].mean()
    summary.append({'genre': col.replace('genre_',''), 'pct_1hw': pct_1hw, 'pct_hitmaker': pct_hm})

summary_df = pd.DataFrame(summary).sort_values('pct_hitmaker', ascending=False)
print(summary_df.to_string(index=False))
```

```python
# 1. Handle genre missingness: add genre_unknown flag, fill genre columns with 0
df_ml_sandbox['genre_unknown'] = df_ml_sandbox['combined_major_genre_categories'].isna().astype(int)

genre_cols = [c for c in df_ml_sandbox.columns if c.startswith('genre_') and c != 'genre_unknown']
for col in genre_cols:
    df_ml_sandbox[col] = df_ml_sandbox[col].fillna(0).astype(int)

print(f"genre_unknown: {df_ml_sandbox['genre_unknown'].sum()} artists ({df_ml_sandbox['genre_unknown'].mean():.1%})")
print(f"Genre columns null rate after fill: {df_ml_sandbox[genre_cols].isnull().sum().sum()} nulls remaining")

# 2. Check songs_through_first_top_10_hit
print(f"\n── songs_through_first_top_10_hit ──")
print(f"Null rate: {df_ml_sandbox['songs_through_first_top_10_hit'].isna().mean():.1%}")
print(f"Null count: {df_ml_sandbox['songs_through_first_top_10_hit'].isna().sum()}")

for label, grp in [('1-Hit Wonders', df_ml_sandbox[df_ml_sandbox['hitmaker']==0]),
                   ('Hitmakers',     df_ml_sandbox[df_ml_sandbox['hitmaker']==1])]:
    col = grp['songs_through_first_top_10_hit']
    print(f"\n  {label} (n={len(grp)})")
    print(f"    min={col.min():.0f}  median={col.median():.0f}  "
          f"mean={col.mean():.2f}  max={col.max():.0f}  nulls={col.isna().sum()}")
    print(f"    Value counts (top 10):")
    print(col.value_counts().head(10).to_string())
```

```python
# ── ML Pipeline: 4 models + baseline, stratified 5-fold CV ────────────────────
from sklearn.linear_model    import LogisticRegression
from sklearn.ensemble        import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm             import SVC
from sklearn.dummy           import DummyClassifier
from sklearn.preprocessing   import StandardScaler
from sklearn.pipeline        import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_validate, cross_val_predict
from sklearn.metrics         import roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# ── CV strategy ───────────────────────────────────────────────────────────────
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# ── Scoring ───────────────────────────────────────────────────────────────────
scoring = {
    'roc_auc':   'roc_auc',
    'f1':        'f1',
    'precision': 'precision',
    'recall':    'recall',
    'accuracy':  'accuracy',
}

# ── Models ────────────────────────────────────────────────────────────────────
models = {
    'Dummy (baseline)': DummyClassifier(strategy='stratified', random_state=42),
    'Logistic Regression': Pipeline([
        ('scaler', StandardScaler()),
        ('clf', LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42))
    ]),
    'Random Forest': RandomForestClassifier(
        n_estimators=300, class_weight='balanced', random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=200, max_depth=3, random_state=42),
    'SVM': Pipeline([
        ('scaler', StandardScaler()),
        ('clf', SVC(class_weight='balanced', probability=True, random_state=42))
    ]),
}

# ── Run CV on training set only ───────────────────────────────────────────────
results = {}
for name, model in models.items():
    print(f"Fitting {name}...")
    scores = cross_validate(model, X_train, y_train, cv=cv, scoring=scoring,
                            return_train_score=True, n_jobs=-1)
    results[name] = scores

# ── Results table ─────────────────────────────────────────────────────────────
print("\n── 5-Fold CV Results (mean ± std) ──\n")
metrics = [('roc_auc','AUC-ROC'), ('f1','F1'), ('precision','Precision'),
           ('recall','Recall'), ('accuracy','Accuracy')]

rows = []
for model_name, scores in results.items():
    row = {'Model': model_name}
    for key, label in metrics:
        v = scores[f'test_{key}']
        row[label] = f"{v.mean():.3f} ± {v.std():.3f}"
    train_auc = scores['train_roc_auc'].mean()
    test_auc  = scores['test_roc_auc'].mean()
    row['Train AUC'] = f"{train_auc:.3f}" if model_name != 'Dummy (baseline)' else '—'
    row['Gap']       = f"{train_auc - test_auc:.3f}" if model_name != 'Dummy (baseline)' else '—'
    rows.append(row)

results_df = pd.DataFrame(rows).set_index('Model')
print(results_df.iloc[:1].to_string())
print('─' * 95)
print(results_df.iloc[1:].to_string())

# ── Feature importances / coefficients (fit on training data only) ────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Feature Importances / Coefficients (fit on training data)',
             fontsize=13, fontweight='bold')

importance_models = {
    'Logistic Regression': (models['Logistic Regression'], 'coef'),
    'Random Forest':       (models['Random Forest'],       'importance'),
    'Gradient Boosting':   (models['Gradient Boosting'],   'importance'),
}

for ax, (model_name, (model, imp_type)) in zip(axes, importance_models.items()):
    model.fit(X_train, y_train)
    importances = (model.named_steps['clf'].coef_[0] if imp_type == 'coef'
                   else model.feature_importances_)
    idx    = np.argsort(np.abs(importances))[-15:]
    colors = ['#EE7733' if importances[i] >= 0 else '#4477AA' for i in idx]
    ax.barh([feature_cols[i] for i in idx], importances[idx], color=colors)
    ax.axvline(0, color='black', linewidth=0.8)
    ax.set_title(model_name, fontsize=11, fontweight='bold')
    ax.grid(axis='x', alpha=0.3)
    if model_name == 'Logistic Regression':
        ax.set_xlabel('Coefficient (orange = → hitmaker)')

plt.tight_layout()
plt.show()

# ── ROC curves (CV on training set) ──────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot([0,1],[0,1],'k--', alpha=0.4, label='Random chance')

linestyles = {'Dummy (baseline)': ':', 'Logistic Regression': '-',
              'Random Forest': '-', 'Gradient Boosting': '-', 'SVM': '-'}
colors_roc  = {'Dummy (baseline)': 'gray', 'Logistic Regression': '#4477AA',
               'Random Forest': '#228833', 'Gradient Boosting': '#EE7733', 'SVM': '#CC3311'}

for model_name, model in models.items():
    probs = cross_val_predict(model, X_train, y_train, cv=cv,
                              method='predict_proba', n_jobs=-1)[:,1]
    fpr, tpr, _ = roc_curve(y_train, probs)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, linewidth=2,
            linestyle=linestyles[model_name],
            color=colors_roc[model_name],
            label=f'{model_name} (AUC={roc_auc:.3f})')

ax.set_xlabel('False Positive Rate', fontsize=11)
ax.set_ylabel('True Positive Rate', fontsize=11)
ax.set_title('ROC Curves — 5-Fold CV on Training Set', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ── Hyperparameter tuning via RandomizedSearchCV (training set only) ───────────
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint
import pandas as pd

tuning_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# ── Parameter grids ───────────────────────────────────────────────────────────
param_grids = {
    'Logistic Regression': {
        'model': Pipeline([
            ('scaler', StandardScaler()),
            ('clf', LogisticRegression(class_weight='balanced', max_iter=2000, random_state=42))
        ]),
        'params': {
            'clf__C':       [0.001, 0.01, 0.1, 0.5, 1, 5, 10, 100],
            'clf__penalty': ['l1', 'l2'],
            'clf__solver':  ['liblinear'],
        },
        'n_iter': 16,
    },
    'Random Forest': {
        'model': RandomForestClassifier(class_weight='balanced', random_state=42, n_jobs=-1),
        'params': {
            'n_estimators':      [100, 200, 300, 500],
            'max_depth':         [3, 4, 5, 7, 10, None],
            'min_samples_leaf':  [2, 4, 8, 15, 25],
            'min_samples_split': [5, 10, 20],
            'max_features':      ['sqrt', 'log2', 0.5],
        },
        'n_iter': 60,
    },
    'Gradient Boosting': {
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators':     [50, 100, 150, 200],
            'max_depth':        [2, 3, 4],
            'learning_rate':    [0.01, 0.05, 0.1, 0.2],
            'min_samples_leaf': [5, 10, 20, 30],
            'subsample':        [0.7, 0.8, 1.0],
        },
        'n_iter': 60,
    },
    'SVM': {
        'model': Pipeline([
            ('scaler', StandardScaler()),
            ('clf', SVC(class_weight='balanced', probability=True, random_state=42))
        ]),
        'params': {
            'clf__C':      [0.01, 0.1, 0.5, 1, 5, 10, 50],
            'clf__gamma':  ['scale', 'auto', 0.001, 0.01, 0.1],
            'clf__kernel': ['rbf', 'linear'],
        },
        'n_iter': 40,
    },
}

# ── Run tuning ────────────────────────────────────────────────────────────────
best_models   = {}
tuning_results = []

for name, cfg in param_grids.items():
    print(f"Tuning {name}...")
    search = RandomizedSearchCV(
        cfg['model'], cfg['params'],
        n_iter=cfg['n_iter'],
        scoring='roc_auc',
        cv=tuning_cv,
        random_state=42,
        n_jobs=-1,
        return_train_score=True,
    )
    search.fit(X_train, y_train)
    best_models[name] = search.best_estimator_

    tuning_results.append({
        'Model':          name,
        'Best CV AUC':    f"{search.best_score_:.3f}",
        'Best Params':    str(search.best_params_),
    })
    print(f"  Best AUC: {search.best_score_:.3f}  |  {search.best_params_}")

print("\n── Tuning Summary ──")
print(pd.DataFrame(tuning_results).set_index('Model').to_string())

# ── Compare tuned models via full CV ─────────────────────────────────────────
print("\n── Tuned Models: Full CV Evaluation ──\n")
tuned_results = {}
rows = []

for name, model in best_models.items():
    scores = cross_validate(model, X_train, y_train, cv=tuning_cv,
                            scoring=scoring, return_train_score=True, n_jobs=-1)
    tuned_results[name] = scores
    row = {'Model': name}
    for key, label in metrics:
        v = scores[f'test_{key}']
        row[label] = f"{v.mean():.3f} ± {v.std():.3f}"
    train_auc = scores['train_roc_auc'].mean()
    test_auc  = scores['test_roc_auc'].mean()
    row['Train AUC'] = f"{train_auc:.3f}"
    row['Gap']       = f"{train_auc - test_auc:.3f}"
    rows.append(row)

print(pd.DataFrame(rows).set_index('Model').to_string())

# ── ROC curves: tuned models ──────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot([0,1],[0,1],'k--', alpha=0.4, label='Random chance')

colors_roc = {'Logistic Regression': '#4477AA', 'Random Forest': '#228833',
              'Gradient Boosting': '#EE7733', 'SVM': '#CC3311'}

for name, model in best_models.items():
    probs = cross_val_predict(model, X_train, y_train, cv=tuning_cv,
                              method='predict_proba', n_jobs=-1)[:,1]
    fpr, tpr, _ = roc_curve(y_train, probs)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, linewidth=2, color=colors_roc[name],
            label=f'{name} (AUC={roc_auc:.3f})')

ax.set_xlabel('False Positive Rate', fontsize=11)
ax.set_ylabel('True Positive Rate', fontsize=11)
ax.set_title('ROC Curves — Tuned Models, 5-Fold CV on Training Set',
             fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Reconstruct best models using known optimal hyperparameters from tuning,
# fit on full training set, and evaluate both on the held-out test set.

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import (
    roc_auc_score, f1_score, precision_score, recall_score,
    confusion_matrix, classification_report, RocCurveDisplay
)
import matplotlib.pyplot as plt
import numpy as np

# --- Reconstruct best models with known tuned params ---
best_lr = LogisticRegression(
    C=0.1, penalty='l1', solver='liblinear',
    class_weight='balanced', max_iter=1000, random_state=42
)
best_gb = GradientBoostingClassifier(
    learning_rate=0.01, max_depth=2, min_samples_leaf=5,
    n_estimators=150, subsample=0.7, random_state=42
)

models = {
    'Logistic Regression': best_lr,
    'Gradient Boosting':   best_gb,
}

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)

    y_pred      = model.predict(X_test)
    y_prob      = model.predict_proba(X_test)[:, 1]

    auc  = roc_auc_score(y_test, y_prob)
    f1   = f1_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec  = recall_score(y_test, y_pred)
    cm   = confusion_matrix(y_test, y_pred)

    results[name] = dict(auc=auc, f1=f1, precision=prec, recall=rec, cm=cm, y_prob=y_prob)

    print(f"\n{'='*50}")
    print(f"  {name}")
    print(f"{'='*50}")
    print(f"  Test AUC:       {auc:.4f}   (CV AUC: {'0.762' if 'Logistic' in name else '0.769'})")
    print(f"  F1 (hitmaker):  {f1:.4f}")
    print(f"  Precision:      {prec:.4f}")
    print(f"  Recall:         {rec:.4f}")
    print(f"\nConfusion Matrix (rows=actual, cols=predicted):")
    print(f"            Pred 1HW  Pred HM")
    print(f"  Actual 1HW   {cm[0,0]:4d}     {cm[0,1]:4d}")
    print(f"  Actual HM    {cm[1,0]:4d}     {cm[1,1]:4d}")
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['1-hit wonder', 'hitmaker']))

# --- ROC curves ---
ax_roc = axes[2]
for name, res in results.items():
    RocCurveDisplay.from_predictions(
        y_test, res['y_prob'],
        name=f"{name} (AUC={res['auc']:.3f})",
        ax=ax_roc
    )
ax_roc.plot([0,1],[0,1],'k--', label='Random baseline')
ax_roc.set_title('ROC Curves — Test Set')
ax_roc.legend(loc='lower right', fontsize=9)

# --- Confusion matrices ---
for ax, (name, res) in zip(axes[:2], results.items()):
    im = ax.imshow(res['cm'], interpolation='nearest', cmap='Blues')
    ax.set_title(f'{name}\nConfusion Matrix')
    ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
    ax.set_xticks([0, 1]); ax.set_xticklabels(['1HW', 'HM'])
    ax.set_yticks([0, 1]); ax.set_yticklabels(['1HW', 'HM'])
    for i in range(2):
        for j in range(2):
            ax.text(j, i, str(res['cm'][i, j]),
                    ha='center', va='center',
                    color='white' if res['cm'][i,j] > res['cm'].max()/2 else 'black',
                    fontsize=14, fontweight='bold')
    plt.colorbar(im, ax=ax)

plt.tight_layout()
plt.show()

# --- Summary ---
print("\n=== SUMMARY ===")
print(f"{'Model':<25} {'Test AUC':>10} {'F1':>8} {'Precision':>10} {'Recall':>8}")
print("-" * 65)
for name, res in results.items():
    print(f"{name:<25} {res['auc']:>10.4f} {res['f1']:>8.4f} {res['precision']:>10.4f} {res['recall']:>8.4f}")
```
