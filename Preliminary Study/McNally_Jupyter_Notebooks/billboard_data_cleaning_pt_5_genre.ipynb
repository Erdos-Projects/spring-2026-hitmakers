```python
"""
Add #_of_genres column to count distinct major genre categories
"""

import pandas as pd

def count_genres(genre_category_string):
    """
    Count the number of distinct major genre categories.
    
    Examples:
        None → 0
        "Hip Hop/Rap" → 1
        "Country/Americana, Folk, Rock" → 3
    """
    if pd.isna(genre_category_string):
        return 0
    
    # Split by comma and count
    categories = [c.strip() for c in str(genre_category_string).split(',')]
    return len(categories)

# Add to df_artists
df_artists['#_of_genres'] = df_artists['major_genre_category'].apply(count_genres)

# Add to df_songs
df_songs['#_of_genres'] = df_songs['major_genre_category'].apply(count_genres)

# Add to df_albums
df_albums['#_of_genres'] = df_albums['major_genre_category'].apply(count_genres)

# Summary statistics
print("Genre Count Summary:")
print("="*80)

print("\ndf_artists:")
print(df_artists['#_of_genres'].value_counts().sort_index().to_string())
print(f"  Mean genres per artist: {df_artists['#_of_genres'].mean():.2f}")

print("\ndf_songs:")
print(df_songs['#_of_genres'].value_counts().sort_index().to_string())
print(f"  Mean genres per song: {df_songs['#_of_genres'].mean():.2f}")

print("\ndf_albums:")
print(df_albums['#_of_genres'].value_counts().sort_index().to_string())
print(f"  Mean genres per album: {df_albums['#_of_genres'].mean():.2f}")

print("\n" + "="*80)
print("COMPLETE: Added '#_of_genres' column to all three dataframes")
print("="*80)
```

```python
# Updated classification that automatically generates keywords from all major labels

import pandas as pd

def add_label_classification_columns_v4(df, label_sets):
    """
    Add 7 classification columns based on label affiliations
    Automatically extracts keywords from all major labels by removing common suffixes
    """
    df = df.copy()
    
    # Extract keywords from all labels by removing common suffixes
    common_suffixes = [
        ' Records', ' Recordings', ' Recording', ' Record',
        ' Music', ' Music Group', ' Musik',
        ' Entertainment', ' Ent.', ' Ent',
        ' Label', ' Labels', ' Label Group',
        ' Productions', ' Production',
        ' Studios', ' Studio',
        ' Corporation', ' Corp.',
        ' Group', ' LLC', ' Inc.', ' Ltd.', ' Limited',
        ' UK', ' USA', ' America', ' International',
        ' Nashville', ' Japan', ' Korea'
    ]
    
    def extract_keyword(label):
        """Extract base keyword by removing common suffixes"""
        keyword = label
        for suffix in common_suffixes:
            if keyword.endswith(suffix):
                keyword = keyword[:-len(suffix)]
        return keyword.strip()
    
    # Build keyword sets for each major label group
    warner_keywords = {extract_keyword(label).lower() for label in label_sets['warner']}
    sony_keywords = {extract_keyword(label).lower() for label in label_sets['sony']}
    universal_keywords = {extract_keyword(label).lower() for label in label_sets['universal']}
    all_major_keywords = warner_keywords | sony_keywords | universal_keywords
    
    # Also keep the original manual keywords
    warner_keywords.update(['warner', 'atlantic'])
    universal_keywords.update(['universal', 'interscope', 'capitol', 'virgin'])
    sony_keywords.update(['sony', 'columbia', 'rca', 'epic', 'arista'])
    all_major_keywords = warner_keywords | sony_keywords | universal_keywords
    
    print(f"Generated {len(all_major_keywords)} total keywords from major label sets")
    print(f"  Warner: {len(warner_keywords)} keywords")
    print(f"  Sony: {len(sony_keywords)} keywords")
    print(f"  Universal: {len(universal_keywords)} keywords")
    
    # Helper function to parse labels from comma-separated string
    def parse_labels(label_string):
        if pd.isna(label_string) or not isinstance(label_string, str):
            return []
        return [l.strip() for l in label_string.split(',') if l.strip()]
    
    # Helper function to check if label matches any major label
    def is_major_label(label):
        label_lower = label.lower()
        label_keyword = extract_keyword(label).lower()
        
        # 1. Exact match against full label sets
        if label in label_sets['all_major']:
            return True
        
        # 2. Keyword match - check if extracted keyword is in major keywords
        if label_keyword in all_major_keywords and len(label_keyword) >= 3:
            return True
        
        # 3. Check if any major keyword is contained in the label
        for keyword in all_major_keywords:
            if len(keyword) >= 4 and keyword in label_lower:
                return True
        
        return False
    
    # Helper function to check which big 3 a label belongs to
    def check_big3_affiliation(label):
        label_lower = label.lower()
        label_keyword = extract_keyword(label).lower()
        
        is_warner = False
        is_universal = False
        is_sony = False
        
        # Exact match
        if label in label_sets['warner']:
            is_warner = True
        if label in label_sets['universal']:
            is_universal = True
        if label in label_sets['sony']:
            is_sony = True
        
        # Keyword match
        if not is_warner and (label_keyword in warner_keywords or 
                             any(kw in label_lower for kw in warner_keywords if len(kw) >= 4)):
            is_warner = True
        
        if not is_universal and (label_keyword in universal_keywords or 
                                any(kw in label_lower for kw in universal_keywords if len(kw) >= 4)):
            is_universal = True
        
        if not is_sony and (label_keyword in sony_keywords or 
                           any(kw in label_lower for kw in sony_keywords if len(kw) >= 4)):
            is_sony = True
        
        return is_warner, is_universal, is_sony
    
    # Initialize lists to store results
    major_label_bool = []
    major_label_names_list = []
    other_label_bool = []
    other_label_names_list = []
    warner_bool = []
    universal_bool = []
    sony_bool = []
    
    # Process each row
    for label_string in df['label']:
        labels = parse_labels(label_string)
        
        # Classify each label
        major_labels = []
        other_labels = []
        row_is_warner = False
        row_is_universal = False
        row_is_sony = False
        
        for label in labels:
            if is_major_label(label):
                major_labels.append(label)
                w, u, s = check_big3_affiliation(label)
                if w:
                    row_is_warner = True
                if u:
                    row_is_universal = True
                if s:
                    row_is_sony = True
            else:
                other_labels.append(label)
        
        # Append results
        major_label_bool.append(len(major_labels) > 0)
        major_label_names_list.append(', '.join(major_labels) if major_labels else '')
        other_label_bool.append(len(other_labels) > 0)
        other_label_names_list.append(', '.join(other_labels) if other_labels else '')
        warner_bool.append(row_is_warner)
        universal_bool.append(row_is_universal)
        sony_bool.append(row_is_sony)
    
    # Get position of label column
    label_col_idx = df.columns.get_loc('label')
    
    # Create new columns
    new_columns = pd.DataFrame({
        'major_label': major_label_bool,
        'major_label_names': major_label_names_list,
        'other_label': other_label_bool,
        'other_label_names': other_label_names_list,
        'warner': warner_bool,
        'universal': universal_bool,
        'sony': sony_bool
    }, index=df.index)
    
    # Insert new columns after 'label' column
    df_left = df.iloc[:, :label_col_idx+1]
    df_right = df.iloc[:, label_col_idx+1:]
    df = pd.concat([df_left, new_columns, df_right], axis=1)
    
    return df

# Clean up first
print("Removing duplicate columns...")
df_songs = df_songs.loc[:, ~df_songs.columns.duplicated(keep='first')]
df_albums = df_albums.loc[:, ~df_albums.columns.duplicated(keep='first')]

# Remove old columns
cols_to_remove = ['major_label', 'major_label_names', 'other_label', 'other_label_names',
                 'independent_label', 'independent_label_names', 'warner', 'universal', 'sony']
df_songs = df_songs.drop(columns=[c for c in cols_to_remove if c in df_songs.columns])
df_albums = df_albums.drop(columns=[c for c in cols_to_remove if c in df_albums.columns])

# Apply new classification
print("\nApplying label classification with auto-generated keywords...")
df_songs = add_label_classification_columns_v4(df_songs, label_sets)
df_albums = add_label_classification_columns_v4(df_albums, label_sets)

print(f"\n{'='*60}")
print("Results:")
print(f"{'='*60}")
print(f"df_songs: {int(df_songs['major_label'].sum())} major ({df_songs['major_label'].mean()*100:.1f}%)")
print(f"df_albums: {int(df_albums['major_label'].sum())} major ({df_albums['major_label'].mean()*100:.1f}%)")
```

```python
"""
LABEL POWER METRICS

This code creates three dataframes measuring record label performance based on 
their artists' chart success:

POWER METRIC DEFINITION:
The "power" of a chart performance is calculated as: 101 - peak_position

This creates an inverse scale where higher chart positions receive more points:
  - #1 hit  = 100 points (101 - 1)
  - #10 hit = 91 points  (101 - 10)
  - #50 hit = 51 points  (101 - 50)
  - #100 hit = 1 point   (101 - 100)

RATIONALE:
This metric captures that a #1 hit is significantly more valuable than a #50 hit.
Total power for a label aggregates these scores across all their charting artists,
providing a measure of the label's overall chart dominance rather than just 
counting number of artists.

DATAFRAMES CREATED (will overwrite if they exist):

1. df_labels_overall - Aggregate metrics across all time
   - label: Record label name
   - big_3_affiliation: Warner, Sony, Universal, or Other
   - #_of_connected_artists: Total distinct artists who charted on this label
   - total_power_of_connected_artists: Sum of power scores across all artists

2. df_labels_yearly - Year-by-year metrics with rolling windows
   - label: Record label name
   - big_3_affiliation: Warner, Sony, Universal, or Other
   - year: Calendar year
   - #_of_connected_artists: Distinct artists who charted that year
   - #_of_connected_artists_rolling5: Distinct artists in 5-year window
   - total_power_of_connected_artists: Sum of power scores for that year
   - total_power_of_connected_artists_rolling5: Sum for 5-year window

3. df_big3_yearly - Big 3 aggregated metrics by year
   - big_3_affiliation: Warner, Sony, Universal, or Other
   - year: Calendar year
   - #_of_connected_artists: Distinct artists across all labels in group
   - #_of_connected_artists_rolling5: Distinct artists in 5-year window
   - total_power_of_connected_artists: Aggregated power for that year
   - total_power_of_connected_artists_rolling5: Aggregated power for 5-year window

NOTE: When an artist appears multiple times in a year, only their best 
(highest power) chart position is counted to avoid double-counting.
"""

import pandas as pd
import numpy as np
from collections import defaultdict

# Check if overwriting existing dataframes
if 'df_labels_overall' in globals():
    print("⚠️  Overwriting existing df_labels_overall")
if 'df_labels_yearly' in globals():
    print("⚠️  Overwriting existing df_labels_yearly")
if 'df_big3_yearly' in globals():
    print("⚠️  Overwriting existing df_big3_yearly")

print("\nBuilding label metrics dataframes...")

# Define legal/corporate suffixes that are not record labels
NON_LABEL_SUFFIXES = {
    'Inc.', 'Inc', 'LLC', 'Ltd.', 'Ltd', 'Limited', 'Corp.', 'Corp', 
    'Corporation', 'L.L.C.', 'Co.', 'Company', 'LP', 'L.P.'
}

# Function to determine big 3 affiliation
def get_big3_affiliation(label):
    """Determine if label belongs to Warner, Sony, Universal, or Other"""
    if label in warner_labels:
        return 'Warner'
    elif label in sony_labels:
        return 'Sony'
    elif label in universal_labels:
        return 'Universal'
    else:
        return 'Other'

# Step 1: Extract all label-artist-year-position combinations from songs and albums
def extract_label_data(df, source_type):
    """Extract label, artist, year, and peak position data"""
    records = []
    
    # Determine year and position columns
    if source_type == 'song':
        year_col = 'first_charting_year'
        pos_col = 'peak_pos'
        artist_col = 'performer_normalized'
    else:  # album
        # Find year column
        year_col = None
        for col in ['year', 'album_year', 'release_year', 'first_charting_year', 'first_year']:
            if col in df.columns:
                year_col = col
                break
        pos_col = 'peak_pos'
        artist_col = 'album_artist' if 'album_artist' in df.columns else 'performer_normalized'
    
    if not year_col or year_col not in df.columns:
        print(f"Warning: No year column found for {source_type}s")
        return []
    
    for idx, row in df.iterrows():
        labels_str = row.get('label', '')
        artist = row.get(artist_col, '')
        year = row.get(year_col, None)
        peak_pos = row.get(pos_col, None)
        
        # Skip if missing critical data
        if pd.isna(labels_str) or pd.isna(artist) or pd.isna(year) or pd.isna(peak_pos):
            continue
        
        # Parse comma-separated labels and filter out legal suffixes
        labels = [l.strip() for l in str(labels_str).split(',') if l.strip()]
        labels = [l for l in labels if l not in NON_LABEL_SUFFIXES]
        
        for label in labels:
            records.append({
                'label': label,
                'artist': artist,
                'year': int(year),
                'peak_pos': int(peak_pos),
                'source': source_type
            })
    
    return records

# Extract data from both dataframes
print("Extracting label data from songs...")
song_records = extract_label_data(df_songs, 'song')
print(f"  Found {len(song_records)} song-label associations")

print("Extracting label data from albums...")
album_records = extract_label_data(df_albums, 'album')
print(f"  Found {len(album_records)} album-label associations")

# Combine all records
all_records = song_records + album_records
df_label_records = pd.DataFrame(all_records)

if len(df_label_records) == 0:
    print("Error: No label data found!")
else:
    print(f"\nTotal label-artist-year records: {len(df_label_records)}")
    
    # Calculate power metric: 101 - peak_pos
    df_label_records['power'] = 101 - df_label_records['peak_pos']
    
    # Add big_3_affiliation to label records
    df_label_records['big_3_affiliation'] = df_label_records['label'].apply(get_big3_affiliation)
    
    # ===================================================================
    # df_labels_overall: Aggregate metrics across all years
    # ===================================================================
    print("\nBuilding df_labels_overall...")
    
    overall_metrics = []
    
    for label in df_label_records['label'].unique():
        label_data = df_label_records[df_label_records['label'] == label]
        
        # Number of distinct artists (across all years)
        num_artists = label_data['artist'].nunique()
        
        # Total power (sum across all artist-year combinations)
        # Use the best (max) peak position per artist per year
        best_per_artist_year = label_data.groupby(['artist', 'year'])['power'].max().sum()
        
        # Determine big 3 affiliation
        big3 = get_big3_affiliation(label)
        
        overall_metrics.append({
            'label': label,
            'big_3_affiliation': big3,
            '#_of_connected_artists': num_artists,
            'total_power_of_connected_artists': best_per_artist_year
        })
    
    df_labels_overall = pd.DataFrame(overall_metrics)
    df_labels_overall = df_labels_overall.sort_values('total_power_of_connected_artists', ascending=False)
    
    print(f"  Created df_labels_overall with {len(df_labels_overall)} labels")
    print(f"    Warner: {(df_labels_overall['big_3_affiliation'] == 'Warner').sum()}")
    print(f"    Sony: {(df_labels_overall['big_3_affiliation'] == 'Sony').sum()}")
    print(f"    Universal: {(df_labels_overall['big_3_affiliation'] == 'Universal').sum()}")
    print(f"    Other: {(df_labels_overall['big_3_affiliation'] == 'Other').sum()}")
    
    # ===================================================================
    # df_labels_yearly: Year-by-year metrics with rolling windows
    # ===================================================================
    print("\nBuilding df_labels_yearly...")
    
    yearly_metrics = []
    
    for label in df_label_records['label'].unique():
        label_data = df_label_records[df_label_records['label'] == label]
        big3 = get_big3_affiliation(label)
        
        # Get all years this label had activity
        years = sorted(label_data['year'].unique())
        
        for year in years:
            # Single year metrics
            year_data = label_data[label_data['year'] == year]
            num_artists_year = year_data['artist'].nunique()
            power_year = year_data.groupby('artist')['power'].max().sum()
            
            # Rolling 5-year window
            rolling_data = label_data[
                (label_data['year'] >= year - 4) & 
                (label_data['year'] <= year)
            ]
            num_artists_rolling5 = rolling_data['artist'].nunique()
            power_rolling5 = rolling_data.groupby(['artist', 'year'])['power'].max().sum()
            
            yearly_metrics.append({
                'label': label,
                'big_3_affiliation': big3,
                'year': year,
                '#_of_connected_artists': num_artists_year,
                '#_of_connected_artists_rolling5': num_artists_rolling5,
                'total_power_of_connected_artists': power_year,
                'total_power_of_connected_artists_rolling5': power_rolling5
            })
    
    df_labels_yearly = pd.DataFrame(yearly_metrics)
    df_labels_yearly = df_labels_yearly.sort_values(['label', 'year'])
    
    print(f"  Created df_labels_yearly with {len(df_labels_yearly)} label-year combinations")
    
    # ===================================================================
    # df_big3_yearly: Aggregated Big 3 metrics by year
    # ===================================================================
    print("\nBuilding df_big3_yearly...")
    
    big3_yearly_metrics = []
    
    # Get all unique big3-year combinations
    for big3 in ['Warner', 'Sony', 'Universal', 'Other']:
        big3_data = df_label_records[df_label_records['big_3_affiliation'] == big3]
        
        if len(big3_data) == 0:
            continue
        
        years = sorted(big3_data['year'].unique())
        
        for year in years:
            # Single year metrics - count DISTINCT artists across all labels in this big3
            year_data = big3_data[big3_data['year'] == year]
            num_artists_year = year_data['artist'].nunique()
            
            # Best power per artist in this year (across all labels in this big3)
            power_year = year_data.groupby('artist')['power'].max().sum()
            
            # Rolling 5-year window
            rolling_data = big3_data[
                (big3_data['year'] >= year - 4) & 
                (big3_data['year'] <= year)
            ]
            num_artists_rolling5 = rolling_data['artist'].nunique()
            power_rolling5 = rolling_data.groupby(['artist', 'year'])['power'].max().sum()
            
            big3_yearly_metrics.append({
                'big_3_affiliation': big3,
                'year': year,
                '#_of_connected_artists': num_artists_year,
                '#_of_connected_artists_rolling5': num_artists_rolling5,
                'total_power_of_connected_artists': power_year,
                'total_power_of_connected_artists_rolling5': power_rolling5
            })
    
    df_big3_yearly = pd.DataFrame(big3_yearly_metrics)
    df_big3_yearly = df_big3_yearly.sort_values(['year', 'big_3_affiliation'])
    
    print(f"  Created df_big3_yearly with {len(df_big3_yearly)} big3-year combinations")
    
    # ===================================================================
    # Display samples
    # ===================================================================
    print(f"\n{'='*80}")
    print("SAMPLE: df_labels_overall (top 10 by power)")
    print(f"{'='*80}")
    print(df_labels_overall[['label', 'big_3_affiliation', '#_of_connected_artists', 'total_power_of_connected_artists']].head(10))
    
    print(f"\n{'='*80}")
    print("SAMPLE: df_labels_yearly (first 10 rows)")
    print(f"{'='*80}")
    print(df_labels_yearly.head(10))
    
    print(f"\n{'='*80}")
    print("SAMPLE: df_big3_yearly (most recent 5 years)")
    print(f"{'='*80}")
    if len(df_big3_yearly) > 0:
        recent_years = df_big3_yearly.nlargest(20, 'year')
        print(recent_years)
    
    print(f"\n{'='*80}")
    print("BIG 3 COMPARISON (overall totals)")
    print(f"{'='*80}")
    big3_summary = df_labels_overall.groupby('big_3_affiliation').agg({
        '#_of_connected_artists': 'sum',
        'total_power_of_connected_artists': 'sum'
    }).sort_values('total_power_of_connected_artists', ascending=False)
    print(big3_summary)
    
    print(f"\n{'='*80}")
    print("SUMMARY")
    print(f"{'='*80}")
    print(f"df_labels_overall: {len(df_labels_overall)} unique labels")
    print(f"df_labels_yearly:  {len(df_labels_yearly)} label-year combinations")
    print(f"df_big3_yearly:    {len(df_big3_yearly)} big3-year combinations")
    if len(df_big3_yearly) > 0:
        print(f"  Year range: {df_big3_yearly['year'].min()} - {df_big3_yearly['year'].max()}")
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# ── 1. Load edge list ─────────────────────────────────────────────────────────
master_edges = pd.read_parquet(
    '/Users/jamesemcnally/Documents/GitHub/spring-2026-hitmakers/master_edge_list_all_artists.parquet'
)
print(f"Loaded {len(master_edges):,} edges")

# ── 2. Build bidirectional adjacency lookup ───────────────────────────────────
edges_fwd = master_edges[['artist_a', 'artist_b', 'first_year']].copy()
edges_fwd.columns = ['target', 'neighbor', 'first_year']
edges_rev = master_edges[['artist_b', 'artist_a', 'first_year']].copy()
edges_rev.columns = ['target', 'neighbor', 'first_year']
edges_both = pd.concat([edges_fwd, edges_rev], ignore_index=True)

adj_grouped = {
    target: grp[['neighbor', 'first_year']].values
    for target, grp in edges_both.groupby('target')
}
print(f"Adjacency dict built for {len(adj_grouped):,} artists")

# ── 3. Build cumulative best peak lookups ─────────────────────────────────────
def build_peak_lookup(df, id_col='musicbrainz_artist_id',
                      year_col='first_charting_year', peak_col='peak_pos'):
    """Returns dict: {artist_id -> array of [year, cumulative_best_peak]}"""
    valid = df[df[id_col].notna()].copy()
    valid[id_col] = valid[id_col].astype(int)

    yearly = (
        valid.groupby([id_col, year_col])[peak_col]
        .min()
        .reset_index()
        .sort_values([id_col, year_col])
    )
    yearly['cumulative_best'] = yearly.groupby(id_col)[peak_col].cummin()

    lookup = {}
    for aid, grp in yearly.groupby(id_col):
        lookup[int(aid)] = grp[[year_col, 'cumulative_best']].values  # numpy for speed
    return lookup

print("Building song peak lookup...")
song_peak_lookup = build_peak_lookup(df_songs)
print(f"  {len(song_peak_lookup):,} artists with song data")

print("Building album peak lookup...")
album_peak_lookup = build_peak_lookup(df_albums)
print(f"  {len(album_peak_lookup):,} artists with album data")

# ── 4. Helper: best peak for an artist as of a given year ─────────────────────
def get_best_peak_as_of(lookup, artist_id, target_year):
    """Returns cumulative best peak position up to target_year, or NaN."""
    if artist_id not in lookup:
        return np.nan
    arr = lookup[artist_id]             # [[year, cumulative_best], ...]
    mask = arr[:, 0] <= target_year
    if not mask.any():
        return np.nan
    return arr[mask, 1][-1]             # last row where year <= target_year

# ── 5. Compute power for one (artist, year) pair ──────────────────────────────
def compute_power(target_artist_id, target_year):
    """Sum of best-of-(song, album) power for all connected artists up to target_year."""
    if pd.isna(target_artist_id) or pd.isna(target_year):
        return np.nan

    target_id  = int(target_artist_id)
    target_year = int(target_year)

    if target_id not in adj_grouped:
        return np.nan

    # Get neighbors connected by target_year
    neighbors = adj_grouped[target_id]
    valid_neighbors = np.unique(neighbors[neighbors[:, 1] <= target_year, 0])

    if len(valid_neighbors) == 0:
        return np.nan

    total_power = 0.0
    for nid in valid_neighbors:
        nid = int(nid)

        # Song power
        song_peak = get_best_peak_as_of(song_peak_lookup, nid, target_year)
        song_power = (101 - song_peak) if not np.isnan(song_peak) else np.nan

        # Album power (normalized to same 0-100 scale)
        album_peak = get_best_peak_as_of(album_peak_lookup, nid, target_year)
        album_power = ((201 - album_peak) / 2) if not np.isnan(album_peak) else np.nan

        # Best of both
        candidates = [p for p in [song_power, album_power] if not np.isnan(p)]
        if candidates:
            total_power += max(candidates)
        # neighbors with no chart history contribute 0

    return total_power if total_power > 0 else np.nan

# ── 6. Apply to all artists ───────────────────────────────────────────────────
power_first_song  = []
power_first_top10 = []

for _, row in tqdm(df_artists.iterrows(), total=len(df_artists), desc="Computing network power"):
    aid        = row['musicbrainz_artist_id']
    song_year  = row['first_song_year']
    top10_year = row['first_year_top_10_songs'] if pd.notna(row.get('first_year_top_10_songs')) else np.nan

    power_first_song.append(compute_power(aid, song_year))
    power_first_top10.append(compute_power(aid, top10_year))

# ── 7. Add columns to df_artists ─────────────────────────────────────────────
new_cols = [
    'power_of_connected_artists_at_first_song_year',
    'power_of_connected_artists_at_first_top10_year'
]
df_artists = df_artists.drop(columns=[c for c in new_cols if c in df_artists.columns])

df_artists['power_of_connected_artists_at_first_song_year']  = power_first_song
df_artists['power_of_connected_artists_at_first_top10_year'] = power_first_top10

# ── 8. Spot check ─────────────────────────────────────────────────────────────
print(f"\ndf_artists shape: {df_artists.shape}")
print(f"\nNon-null values:")
print(f"  power_at_first_song_year:  {df_artists['power_of_connected_artists_at_first_song_year'].notna().sum():,}")
print(f"  power_at_first_top10_year: {df_artists['power_of_connected_artists_at_first_top10_year'].notna().sum():,}")
print("\nSample:")
print(df_artists[['performer_normalized'] + new_cols].dropna(subset=[new_cols[0]]).head(10))
```

```python
from scipy import stats
import numpy as np

periods = [
    (2001, 2005, '2001-2005'),
    (2006, 2010, '2006-2010'),
    (2011, 2015, '2011-2015'),
    (2016, 2020, '2016-2020'),
]

# Paired for direct comparison (yearly vs rolling10)
metric_pairs = [
    ('eigenvector_centrality_top10_yearly',        'eigenvector_centrality_top10_rolling10',        'Eigenvector'),
    ('closeness_centrality_top10_yearly',           'closeness_centrality_top10_rolling10',           'Closeness'),
    ('harmonic_closeness_centrality_top10_yearly',  'harmonic_closeness_centrality_top10_rolling10',  'Harmonic Close.'),
    ('betweenness_centrality_top10_yearly',         'betweenness_centrality_top10_rolling10',         'Betweenness'),
    ('degree_centrality_top10_yearly',              'degree_centrality_top10_rolling10',              'Degree'),
]

def run_test(ohw, hm, metric):
    a = ohw[metric].dropna()
    b = hm[metric].dropna()
    if len(a) < 5 or len(b) < 5:
        return None
    t, p = stats.ttest_ind(a, b, equal_var=False)
    pooled = np.sqrt(((len(a)-1)*a.std()**2 + (len(b)-1)*b.std()**2) / (len(a)+len(b)-2))
    d = (b.mean() - a.mean()) / pooled if pooled > 0 else 0
    sig = '***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else '†' if p < 0.10 else ''
    return a.mean(), b.mean(), t, p, d, sig

for start, end, label in periods:
    df_per = df_top10[
        (df_top10['first_year_top_10_songs'] >= start) &
        (df_top10['first_year_top_10_songs'] <= end)
    ]
    ohw_d = df_per[df_per['top_10_1hit_wonder_song'] == True]
    hm_d  = df_per[df_per['top_10_1hit_wonder_song'] == False]

    print(f"\n{'='*100}")
    print(f"{label}  |  N={len(df_per)} (OHW: {len(ohw_d)}, HM: {len(hm_d)})")
    print(f"{'='*100}")
    print(f"  {'Metric':<18} {'--- YEARLY ---':^38} {'--- ROLLING 10 ---':^38}")
    print(f"  {'':18} {'OHW':>8} {'HM':>8} {'t':>6} {'p':>8} {'d':>6}{'':4} {'OHW':>8} {'HM':>8} {'t':>6} {'p':>8} {'d':>6}")
    print(f"  {'-'*94}")

    for m_yr, m_r10, name in metric_pairs:
        yr  = run_test(ohw_d, hm_d, m_yr)
        r10 = run_test(ohw_d, hm_d, m_r10)

        yr_str  = f"{yr[0]:>8.4f} {yr[1]:>8.4f} {yr[2]:>6.2f} {yr[3]:>8.4f} {yr[4]:>6.3f}{yr[5]:<3}" if yr  else f"{'insufficient data':>42}"
        r10_str = f"{r10[0]:>8.4f} {r10[1]:>8.4f} {r10[2]:>6.2f} {r10[3]:>8.4f} {r10[4]:>6.3f}{r10[5]:<3}" if r10 else f"{'insufficient data':>42}"

        print(f"  {name:<18} {yr_str}    {r10_str}")

    # Power (yearly only)
    print(f"\n  {'Power (yearly only)':<18}", end=' ')
    res = run_test(ohw_d, hm_d, 'power_of_connected_artists_at_first_top10_year')
    if res:
        print(f"{res[0]:>8.4f} {res[1]:>8.4f} {res[2]:>6.2f} {res[3]:>8.4f} {res[4]:>6.3f}{res[5]:<3}")
    else:
        print("insufficient data")
```

```python
# Statistical analysis: Label Power Metrics vs One-Hit Wonder Status (post-2000, top 10 songs)

import pandas as pd
import numpy as np
from scipy import stats

print("="*80)
print("LABEL POWER METRICS vs ONE-HIT WONDER STATUS")
print("Post-2000 Artists with Top 10 Songs")
print("="*80)

# Filter: post-2000 artists with top 10 songs
df_analysis = df_artists[
    (df_artists['first_song_year'] >= 2000) & 
    (df_artists['top_10_song_count'] > 0)
].copy()

# Filter to one-hit wonders and hitmakers only
df_comp = df_analysis[
    (df_analysis['top_10_1hit_wonder_song'] == True) | 
    (df_analysis['top_10_hitmaker_songs'] == True)
].copy()

df_comp['status'] = df_comp['top_10_1hit_wonder_song'].map({True: 'One-Hit Wonder', False: 'Hitmaker'})

print(f"\nSample: {len(df_comp)} artists")
print(f"  One-Hit Wonders: {(df_comp['status']=='One-Hit Wonder').sum()}")
print(f"  Hitmakers: {(df_comp['status']=='Hitmaker').sum()}")

# Metrics to analyze
metrics = [
    'first_top_10_song_#_of_connected_artists_to_labels_single_year',
    'first_top_10_song_#_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_labels_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_#_of_connected_artists_to_big3_single_year',
    'first_top_10_song_#_of_connected_artists_to_big3_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_big3_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_big3_past_5_years'
]

# Short labels for display
metric_labels = [
    'Artists to Labels (1 year)',
    'Artists to Labels (5 years)',
    'Power to Labels (1 year)',
    'Power to Labels (5 years)',
    'Artists to Big 3 (1 year)',
    'Artists to Big 3 (5 years)',
    'Power to Big 3 (1 year)',
    'Power to Big 3 (5 years)'
]

print(f"\n{'-'*80}")
print("RESULTS")
print(f"{'-'*80}\n")

# Analyze each metric
for metric, label in zip(metrics, metric_labels):
    if metric not in df_comp.columns:
        print(f"{label}: Column not found")
        continue
    
    ohw = df_comp[df_comp['status'] == 'One-Hit Wonder'][metric].dropna()
    hm = df_comp[df_comp['status'] == 'Hitmaker'][metric].dropna()
    
    if len(ohw) > 0 and len(hm) > 0:
        t_stat, p_val = stats.ttest_ind(ohw, hm)
        cohens_d = (ohw.mean() - hm.mean()) / np.sqrt(
            ((len(ohw)-1)*ohw.std()**2 + (len(hm)-1)*hm.std()**2) / (len(ohw)+len(hm)-2)
        )
        
        # Determine significance marker
        sig = "***" if p_val < 0.001 else "**" if p_val < 0.01 else "*" if p_val < 0.05 else "ns"
        
        print(f"{label}:")
        print(f"  OHW: M={ohw.mean():.1f}, SD={ohw.std():.1f}, n={len(ohw)}")
        print(f"  HM:  M={hm.mean():.1f}, SD={hm.std():.1f}, n={len(hm)}")
        print(f"  t={t_stat:.2f}, p={p_val:.4f} {sig}, d={cohens_d:.3f}\n")
    else:
        print(f"{label}: Insufficient data\n")

print(f"{'-'*80}")
print("Significance: *** p<.001, ** p<.01, * p<.05, ns = not significant")
print(f"{'='*80}")
```

```python
from scipy.stats import chi2_contingency, ttest_ind
import scipy.stats as stats
import numpy as np

# First, recreate the comparison dataset
df_comp = df_artists[
    (df_artists['first_year_top_10_songs'] >= 2000) & 
    (df_artists['top_10_song_count'] > 0)
].copy()

# Create the binary target variable (using existing columns)
df_comp['is_one_hit_wonder'] = df_comp['top_10_1hit_wonder_song']

print(f"Total artists (post-2000 with top 10 hits): {len(df_comp)}")
print(f"One-Hit Wonders: {df_comp['is_one_hit_wonder'].sum()}")
print(f"Hitmakers: {(~df_comp['is_one_hit_wonder']).sum()}")
print("\n" + "="*80 + "\n")

# Check who has label data
df_comp['has_label_data'] = df_comp['first_top_10_song_labels'].notna() & (df_comp['first_top_10_song_labels'] != '')

# Filter to only artists with label data
df_with_label_data = df_comp[df_comp['has_label_data']].copy()

print(f"Sample size WITH label data: {len(df_with_label_data)}")
print(f"One-Hit Wonders: {df_with_label_data['is_one_hit_wonder'].sum()}")
print(f"Hitmakers: {(~df_with_label_data['is_one_hit_wonder']).sum()}")
print("\n" + "="*80 + "\n")

# 1. Big 3 Affiliation (Yes/No)
print("1. BIG 3 AFFILIATION (YES/NO)")
print("-" * 40)

crosstab_big3 = pd.crosstab(
    df_with_label_data['first_top_10_song_big3_label'].notna() & (df_with_label_data['first_top_10_song_big3_label'] != ''),
    df_with_label_data['is_one_hit_wonder']
)
crosstab_big3.index = crosstab_big3.index.map({True: 'Big 3', False: 'Non-Big 3'})
crosstab_big3.columns = crosstab_big3.columns.map({True: 'One-Hit Wonder', False: 'Hitmaker'})

# Add percentages
crosstab_pct = crosstab_big3.div(crosstab_big3.sum(axis=1), axis=0) * 100
print("\nCounts:")
print(crosstab_big3)
print("\nPercentages (% who are Hitmakers):")
for idx in crosstab_pct.index:
    hitmaker_pct = crosstab_pct.loc[idx, 'Hitmaker']
    print(f"{idx}: {hitmaker_pct:.1f}%")

chi2, p_value, dof, expected = chi2_contingency(crosstab_big3)
n = crosstab_big3.sum().sum()
cramers_v = np.sqrt(chi2 / n)
print(f"\nχ² = {chi2:.2f}, p = {p_value:.4f}, Cramér's V = {cramers_v:.3f}")

# Calculate effect size
big3_hitmaker_pct = crosstab_pct.loc['Big 3', 'Hitmaker']
non_big3_hitmaker_pct = crosstab_pct.loc['Non-Big 3', 'Hitmaker']
print(f"Effect size: {big3_hitmaker_pct - non_big3_hitmaker_pct:.1f} percentage points")

print("\n" + "="*80 + "\n")

# 2. Label Power Metrics
print("2. LABEL POWER METRICS")
print("-" * 40)

metrics = [
    'first_top_10_song_#_of_connected_artists_to_labels_single_year',
    'first_top_10_song_#_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_labels_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_#_of_connected_artists_to_big3_single_year',
    'first_top_10_song_#_of_connected_artists_to_big3_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_big3_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_big3_past_5_years'
]

for metric in metrics:
    # Skip if all NaN
    if df_with_label_data[metric].isna().all():
        continue
    
    one_hit_wonders = df_with_label_data[df_with_label_data['is_one_hit_wonder']][metric].dropna()
    hitmakers = df_with_label_data[~df_with_label_data['is_one_hit_wonder']][metric].dropna()
    
    if len(one_hit_wonders) < 2 or len(hitmakers) < 2:
        continue
    
    mean_ohw = one_hit_wonders.mean()
    mean_hm = hitmakers.mean()
    
    t_stat, p_value = stats.ttest_ind(one_hit_wonders, hitmakers, equal_var=False)
    
    pooled_std = np.sqrt(((len(one_hit_wonders)-1)*one_hit_wonders.std()**2 + 
                          (len(hitmakers)-1)*hitmakers.std()**2) / 
                         (len(one_hit_wonders) + len(hitmakers) - 2))
    cohens_d = (mean_hm - mean_ohw) / pooled_std
    
    print(f"\n{metric}:")
    print(f"  One-Hit Wonders: M={mean_ohw:.1f}, N={len(one_hit_wonders)}")
    print(f"  Hitmakers: M={mean_hm:.1f}, N={len(hitmakers)}")
    print(f"  t={t_stat:.2f}, p={p_value:.4f}, d={cohens_d:.3f}")

print("\n" + "="*80)
```

```python
from scipy.stats import chi2_contingency, ttest_ind
import scipy.stats as stats
import numpy as np
import pandas as pd

# Define 4-year periods
periods = [
    (2001, 2004),
    (2005, 2008),
    (2009, 2012),
    (2013, 2016),
    (2017, 2020)
]

# Store results
results_summary = []

# Label power metrics to test
metrics = [
    'first_top_10_song_#_of_connected_artists_to_labels_single_year',
    'first_top_10_song_#_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_labels_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_labels_past_5_years',
    'first_top_10_song_#_of_connected_artists_to_big3_single_year',
    'first_top_10_song_#_of_connected_artists_to_big3_past_5_years',
    'first_top_10_song_total_power_of_connected_artists_to_big3_single_year',
    'first_top_10_song_total_power_of_connected_artists_to_big3_past_5_years'
]

# Recreate df_comp with label data filter
df_comp = df_artists[
    (df_artists['first_year_top_10_songs'] >= 2000) & 
    (df_artists['top_10_song_count'] > 0)
].copy()
df_comp['is_one_hit_wonder'] = df_comp['top_10_1hit_wonder_song']
df_comp['has_label_data'] = df_comp['first_top_10_song_labels'].notna() & (df_comp['first_top_10_song_labels'] != '')
df_with_label_data = df_comp[df_comp['has_label_data']].copy()

# Loop through periods
for start_year, end_year in periods:
    period_label = f"{start_year}-{end_year}"
    print("="*80)
    print(f"PERIOD: {period_label}")
    print("="*80)
    
    # Filter to period
    df_period = df_with_label_data[
        (df_with_label_data['first_year_top_10_songs'] >= start_year) &
        (df_with_label_data['first_year_top_10_songs'] <= end_year)
    ].copy()
    
    n_total = len(df_period)
    n_ohw = df_period['is_one_hit_wonder'].sum()
    n_hm = (~df_period['is_one_hit_wonder']).sum()
    
    print(f"\nSample size: {n_total}")
    print(f"  One-Hit Wonders: {n_ohw}")
    print(f"  Hitmakers: {n_hm}")
    
    if n_total < 20:  # Skip if too few observations
        print("  ⚠️  Sample too small, skipping period\n")
        continue
    
    print("\n" + "-"*80)
    print("1. BIG 3 AFFILIATION")
    print("-"*80)
    
    # Big 3 affiliation test
    df_period['has_big3'] = df_period['first_top_10_song_big3_label'].notna() & (df_period['first_top_10_song_big3_label'] != '')
    
    crosstab = pd.crosstab(df_period['has_big3'], df_period['is_one_hit_wonder'])
    
    if crosstab.shape == (2, 2):  # Only if we have both categories
        crosstab.index = crosstab.index.map({True: 'Big 3', False: 'Non-Big 3'})
        crosstab.columns = crosstab.columns.map({True: 'One-Hit Wonder', False: 'Hitmaker'})
        
        # Calculate percentages
        crosstab_pct = crosstab.div(crosstab.sum(axis=1), axis=0) * 100
        
        print("\nCounts:")
        print(crosstab)
        
        big3_hitmaker_pct = crosstab_pct.loc['Big 3', 'Hitmaker']
        non_big3_hitmaker_pct = crosstab_pct.loc['Non-Big 3', 'Hitmaker']
        effect_size_pct = big3_hitmaker_pct - non_big3_hitmaker_pct
        
        print(f"\nBig 3 Hitmaker %: {big3_hitmaker_pct:.1f}%")
        print(f"Non-Big 3 Hitmaker %: {non_big3_hitmaker_pct:.1f}%")
        print(f"Effect size: {effect_size_pct:.1f} percentage points")
        
        # Chi-square test
        chi2, p_value, dof, expected = chi2_contingency(crosstab)
        n = crosstab.sum().sum()
        cramers_v = np.sqrt(chi2 / n)
        
        print(f"χ² = {chi2:.2f}, p = {p_value:.4f}, Cramér's V = {cramers_v:.3f}")
        
        # Store result
        results_summary.append({
            'Period': period_label,
            'Metric': 'Big 3 Affiliation',
            'N_Total': n_total,
            'N_OHW': n_ohw,
            'N_HM': n_hm,
            'Effect_Type': 'Percentage Points',
            'Effect_Size': effect_size_pct,
            'Statistic': f"χ²={chi2:.2f}",
            'p_value': p_value,
            'Effect_Measure': f"V={cramers_v:.3f}"
        })
    else:
        print("  ⚠️  Insufficient variation in Big 3 affiliation")
    
    print("\n" + "-"*80)
    print("2. LABEL POWER METRICS")
    print("-"*80)
    
    # Test each metric
    for metric in metrics:
        # Skip if all NaN
        if df_period[metric].isna().all():
            continue
        
        one_hit_wonders = df_period[df_period['is_one_hit_wonder']][metric].dropna()
        hitmakers = df_period[~df_period['is_one_hit_wonder']][metric].dropna()
        
        if len(one_hit_wonders) < 2 or len(hitmakers) < 2:
            continue
        
        mean_ohw = one_hit_wonders.mean()
        mean_hm = hitmakers.mean()
        
        t_stat, p_value = stats.ttest_ind(one_hit_wonders, hitmakers, equal_var=False)
        
        # Cohen's d
        pooled_std = np.sqrt(((len(one_hit_wonders)-1)*one_hit_wonders.std()**2 + 
                              (len(hitmakers)-1)*hitmakers.std()**2) / 
                             (len(one_hit_wonders) + len(hitmakers) - 2))
        
        if pooled_std > 0:
            cohens_d = (mean_hm - mean_ohw) / pooled_std
        else:
            cohens_d = 0
        
        # Shorten metric name for display
        metric_short = metric.replace('first_top_10_song_', '').replace('_of_connected_artists_to_', '_').replace('total_power', 'power')
        
        print(f"\n{metric_short}:")
        print(f"  OHW: M={mean_ohw:.1f} (N={len(one_hit_wonders)}), HM: M={mean_hm:.1f} (N={len(hitmakers)})")
        print(f"  t={t_stat:.2f}, p={p_value:.4f}, d={cohens_d:.3f}")
        
        # Store result
        results_summary.append({
            'Period': period_label,
            'Metric': metric_short,
            'N_Total': n_total,
            'N_OHW': len(one_hit_wonders),
            'N_HM': len(hitmakers),
            'Effect_Type': "Cohen's d",
            'Effect_Size': cohens_d,
            'Statistic': f"t={t_stat:.2f}",
            'p_value': p_value,
            'Effect_Measure': f"d={cohens_d:.3f}"
        })
    
    print("\n")

# Create summary dataframe
df_results = pd.DataFrame(results_summary)
print("\n" + "="*80)
print("SUMMARY TABLE")
print("="*80)
print(df_results.to_string(index=False))
```
