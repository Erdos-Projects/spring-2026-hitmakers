```python
# Critical question: Do one-hit wonders and hitmakers have different NaN rates?
import pandas as pd
import numpy as np

df_artists_top_10_network_metrics = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

metrics = [
    'degree_centrality_top10_yearly',
    'degree_centrality_top10_rolling10',
    'degree_centrality_top10_cumulative',
]

print("="*90)
print("NaN RATE COMPARISON: One-Hit Wonders vs Hitmakers")
print("="*90)

# Overall comparison
print("\nOVERALL (2001-2020):")
print("-"*90)

oww_all = df_artists_top_10_network_metrics[
    (df_artists_top_10_network_metrics['top_10_1hit_wonder_song'] == True) &
    (df_artists_top_10_network_metrics['first_year_top_10_songs'].between(2001, 2020))
]
hm_all = df_artists_top_10_network_metrics[
    (df_artists_top_10_network_metrics['top_10_hitmaker_songs'] == True) &
    (df_artists_top_10_network_metrics['first_year_top_10_songs'].between(2001, 2020))
]

print(f"Total sample: OHW={len(oww_all)}, HM={len(hm_all)}")
print(f"\n{'Metric':<50} {'OHW NaN%':<12} {'HM NaN%':<12} {'Difference':<12}")
print("-"*90)

for metric in metrics:
    oww_nan_pct = oww_all[metric].isna().sum() / len(oww_all) * 100
    hm_nan_pct = hm_all[metric].isna().sum() / len(hm_all) * 100
    diff = oww_nan_pct - hm_nan_pct
    
    print(f"{metric:<50} {oww_nan_pct:>6.1f}%     {hm_nan_pct:>6.1f}%     {diff:>+6.1f}%")

print("\n" + "="*90)
print("INTERPRETATION:")
print("="*90)
print("""
If OHW have HIGHER NaN rates (positive difference):
  â†’ One-hit wonders are LESS likely to be in collaboration networks
  â†’ "Not being collaborative" is associated with one-hit wonder status
  â†’ fillna(0) would STRENGTHEN positive effects (hitmakers have higher centrality)

If HM have HIGHER NaN rates (negative difference):
  â†’ Hitmakers are LESS likely to be in collaboration networks (unlikely)
  â†’ fillna(0) would WEAKEN or reverse effects

If similar NaN rates:
  â†’ Missing data is random with respect to outcome
  â†’ fillna(0) would have minimal directional impact
""")

# Year-by-year comparison for key years
print("\n" + "="*90)
print("YEAR-BY-YEAR NaN RATE COMPARISON (Selected Years)")
print("="*90)

test_years = [2001, 2006, 2010, 2014, 2018]
metric = 'degree_centrality_top10_rolling10'

print(f"\nMetric: {metric}")
print(f"{'Year':<6} {'OHW n':<8} {'OHW NaN%':<12} {'HM n':<8} {'HM NaN%':<12} {'Diff':<10}")
print("-"*90)

for year in test_years:
    df_year = df_artists_top_10_network_metrics[
        df_artists_top_10_network_metrics['first_year_top_10_songs'] == year
    ]
    
    oww = df_year[df_year['top_10_1hit_wonder_song'] == True]
    hm = df_year[df_year['top_10_hitmaker_songs'] == True]
    
    if len(oww) > 0 and len(hm) > 0:
        oww_nan_pct = oww[metric].isna().sum() / len(oww) * 100
        hm_nan_pct = hm[metric].isna().sum() / len(hm) * 100
        diff = oww_nan_pct - hm_nan_pct
        
        print(f"{year:<6} {len(oww):<8} {oww_nan_pct:>6.1f}%     {len(hm):<8} {hm_nan_pct:>6.1f}%     {diff:>+6.1f}%")
```

```python
# Now run decade analysis with fillna(0) to see how it changes results
import pandas as pd
import numpy as np
from scipy import stats

df_artists_top_10_network_metrics = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

def cohens_d(group1, group2):
    """Calculate Cohen's d effect size"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    return (np.mean(group2) - np.mean(group1)) / pooled_std if pooled_std > 0 else 0

decades = [
    (2000, 2009, "2000s"),
    (2010, 2019, "2010s")
]

test_metric = 'degree_centrality_top10_rolling10'

print("="*90)
print(f"COMPARISON: dropna() vs fillna(0) for Decade Analysis")
print(f"Metric: {test_metric}")
print("="*90)

for start_year, end_year, decade_label in decades:
    print(f"\n{'-'*90}")
    print(f"DECADE: {decade_label} ({start_year}-{end_year})")
    print(f"{'-'*90}")
    
    df_decade = df_artists_top_10_network_metrics[
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] >= start_year) &
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] <= end_year)
    ]
    
    oww = df_decade[df_decade['top_10_1hit_wonder_song'] == True]
    hm = df_decade[df_decade['top_10_hitmaker_songs'] == True]
    
    print(f"Total sample: OHW={len(oww)}, HM={len(hm)}")
    
    # Approach 1: dropna
    oww_drop = oww[test_metric].dropna()
    hm_drop = hm[test_metric].dropna()
    
    print(f"\nAPPROACH 1: dropna()")
    print(f"  n: OHW={len(oww_drop)}, HM={len(hm_drop)}")
    print(f"  NaN removed: OHW={len(oww) - len(oww_drop)}, HM={len(hm) - len(hm_drop)}")
    print(f"  Means: OHW={oww_drop.mean():.4f}, HM={hm_drop.mean():.4f}")
    
    if len(oww_drop) >= 2 and len(hm_drop) >= 2:
        _, p1 = stats.ttest_ind(oww_drop, hm_drop, equal_var=False)
        d1 = cohens_d(oww_drop, hm_drop)
        print(f"  Results: p={p1:.4f}, d={d1:.3f}")
    
    # Approach 2: fillna(0)
    oww_fill = oww[test_metric].fillna(0)
    hm_fill = hm[test_metric].fillna(0)
    
    print(f"\nAPPROACH 2: fillna(0)")
    print(f"  n: OHW={len(oww_fill)}, HM={len(hm_fill)}")
    print(f"  Zeros added: OHW={oww[test_metric].isna().sum()}, HM={hm[test_metric].isna().sum()}")
    print(f"  Means: OHW={oww_fill.mean():.4f}, HM={hm_fill.mean():.4f}")
    
    if len(oww_fill) >= 2 and len(hm_fill) >= 2:
        _, p2 = stats.ttest_ind(oww_fill, hm_fill, equal_var=False)
        d2 = cohens_d(oww_fill, hm_fill)
        print(f"  Results: p={p2:.4f}, d={d2:.3f}")
    
    # Show change
    if len(oww_drop) >= 2 and len(hm_drop) >= 2 and len(oww_fill) >= 2 and len(hm_fill) >= 2:
        print(f"\nCHANGE:")
        print(f"  Effect size: {d1:.3f} â†’ {d2:.3f} (Î”={d2-d1:+.3f})")
        print(f"  P-value: {p1:.4f} â†’ {p2:.4f}")
        
        if (p1 < 0.05) != (p2 < 0.05):
            print(f"  âš ï¸  SIGNIFICANCE CHANGED!")
        
        if abs(d2 - d1) > 0.2:
            print(f"  âš ï¸  EFFECT SIZE CHANGED SUBSTANTIALLY!")
```

```python
# Decade-by-decade statistical analysis with fillna(0): Updated metrics
import pandas as pd
import numpy as np
from scipy import stats

df_artists_top_10_network_metrics = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

# Updated metrics: removed cumulative, added eigenvector yearly and rolling10
metrics_to_compare = [
    'degree_centrality_top10_yearly',
    'degree_centrality_top10_rolling10',
    'closeness_centrality_top10_yearly',
    'closeness_centrality_top10_rolling10',
    'harmonic_closeness_centrality_top10_yearly',
    'harmonic_closeness_centrality_top10_rolling10',
    'betweenness_centrality_top10_yearly',
    'betweenness_centrality_top10_rolling10',
    'eigenvector_centrality_top10_yearly',
    'eigenvector_centrality_top10_rolling10'
]

# Define decades
decades = [
    (1960, 1969, "1960s"),
    (1970, 1979, "1970s"),
    (1980, 1989, "1980s"),
    (1990, 1999, "1990s"),
    (2000, 2009, "2000s"),
    (2010, 2019, "2010s")
]

def cohens_d(group1, group2):
    """Calculate Cohen's d effect size"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    return (np.mean(group2) - np.mean(group1)) / pooled_std if pooled_std > 0 else 0

print("="*90)
print("DECADE ANALYSIS WITH fillna(0): Top 10 One-Hit Wonders vs Hitmakers")
print("Updated metrics: Removed cumulative, Added eigenvector (yearly & rolling10)")
print("="*90)

for start_year, end_year, decade_label in decades:
    print(f"\n{'='*90}")
    print(f"DECADE: {decade_label} ({start_year}-{end_year})")
    print(f"{'='*90}")
    
    df_decade = df_artists_top_10_network_metrics[
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] >= start_year) &
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] <= end_year)
    ]
    
    one_hit_wonders = df_decade[df_decade['top_10_1hit_wonder_song'] == True]
    hitmakers = df_decade[df_decade['top_10_hitmaker_songs'] == True]
    
    print(f"\nSample sizes:")
    print(f"  One-Hit Wonders: {len(one_hit_wonders)}")
    print(f"  Hitmakers: {len(hitmakers)}")
    print(f"  Total: {len(df_decade)}")
    
    if len(one_hit_wonders) < 2 or len(hitmakers) < 2:
        print("\n  âš ï¸  Insufficient sample size for statistical testing")
        continue
    
    print(f"\n{'-'*90}")
    print(f"{'Metric':<50} {'t-test p':<12} {'MW p':<12} {'Effect Size':<12}")
    print(f"{'-'*90}")
    
    significant_count_t = 0
    significant_count_mw = 0
    
    for metric in metrics_to_compare:
        # Use fillna(0) instead of dropna()
        oww_values = one_hit_wonders[metric].fillna(0)
        hm_values = hitmakers[metric].fillna(0)
        
        if len(oww_values) < 2 or len(hm_values) < 2:
            print(f"{metric:<50} {'N/A':<12} {'N/A':<12} {'N/A':<12}")
            continue
        
        # Independent t-test
        t_stat, t_pval = stats.ttest_ind(oww_values, hm_values, equal_var=False)
        
        # Mann-Whitney U test
        u_stat, mw_pval = stats.mannwhitneyu(oww_values, hm_values, alternative='two-sided')
        
        # Cohen's d effect size
        effect_size = cohens_d(oww_values, hm_values)
        
        # Format p-values
        t_pval_str = f"{t_pval:.4f}" if t_pval >= 0.0001 else "<0.0001"
        mw_pval_str = f"{mw_pval:.4f}" if mw_pval >= 0.0001 else "<0.0001"
        
        # Add significance markers
        t_sig = " *" if t_pval < 0.05 else ""
        mw_sig = " *" if mw_pval < 0.05 else ""
        
        # Effect size interpretation
        if abs(effect_size) < 0.2:
            effect_interp = "negligible"
        elif abs(effect_size) < 0.5:
            effect_interp = "small"
        elif abs(effect_size) < 0.8:
            effect_interp = "medium"
        else:
            effect_interp = "LARGE"
        
        print(f"{metric:<50} {t_pval_str + t_sig:<12} {mw_pval_str + mw_sig:<12} {effect_size:>6.3f} ({effect_interp})")
        
        if t_pval < 0.05:
            significant_count_t += 1
        if mw_pval < 0.05:
            significant_count_mw += 1
    
    print(f"{'-'*90}")
    print(f"Significant results (p < 0.05): {significant_count_t}/{len(metrics_to_compare)} (t-test), {significant_count_mw}/{len(metrics_to_compare)} (Mann-Whitney)")
    print(f"* = p < 0.05 (statistically significant)")
    print(f"Effect size: Positive = Hitmakers higher centrality, Negative = One-Hit Wonders higher")

print("\n" + "="*90)
print("NOTE: This analysis uses fillna(0), treating missing network data as zero")
print("centrality. One-hit wonders are 16-21% less likely to be in collaboration networks.")
print("="*90)
```

```python
# 2-year period statistical analysis with fillna(0): 2001-2022
import pandas as pd
import numpy as np
from scipy import stats

df_artists_top_10_network_metrics = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

# Updated metrics: removed cumulative, added eigenvector yearly and rolling10
metrics_to_compare = [
    'degree_centrality_top10_yearly',
    'degree_centrality_top10_rolling10',
    'closeness_centrality_top10_yearly',
    'closeness_centrality_top10_rolling10',
    'harmonic_closeness_centrality_top10_yearly',
    'harmonic_closeness_centrality_top10_rolling10',
    'betweenness_centrality_top10_yearly',
    'betweenness_centrality_top10_rolling10',
    'eigenvector_centrality_top10_yearly',
    'eigenvector_centrality_top10_rolling10'
]

# Define 2-year periods from 2001-2022
periods = [
    (2001, 2002, "2001-2002"),
    (2003, 2004, "2003-2004"),
    (2005, 2006, "2005-2006"),
    (2007, 2008, "2007-2008"),
    (2009, 2010, "2009-2010"),
    (2011, 2012, "2011-2012"),
    (2013, 2014, "2013-2014"),
    (2015, 2016, "2015-2016"),
    (2017, 2018, "2017-2018"),
    (2019, 2020, "2019-2020"),
    (2021, 2022, "2021-2022")
]

def cohens_d(group1, group2):
    """Calculate Cohen's d effect size"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    return (np.mean(group2) - np.mean(group1)) / pooled_std if pooled_std > 0 else 0

print("="*90)
print("2-YEAR PERIOD ANALYSIS WITH fillna(0): 2001-2022")
print("Top 10 One-Hit Wonders vs Hitmakers")
print("="*90)

for start_year, end_year, period_label in periods:
    print(f"\n{'='*90}")
    print(f"PERIOD: {period_label}")
    print(f"{'='*90}")
    
    df_period = df_artists_top_10_network_metrics[
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] >= start_year) &
        (df_artists_top_10_network_metrics['first_year_top_10_songs'] <= end_year)
    ]
    
    one_hit_wonders = df_period[df_period['top_10_1hit_wonder_song'] == True]
    hitmakers = df_period[df_period['top_10_hitmaker_songs'] == True]
    
    print(f"\nSample sizes:")
    print(f"  One-Hit Wonders: {len(one_hit_wonders)}")
    print(f"  Hitmakers: {len(hitmakers)}")
    print(f"  Total: {len(df_period)}")
    
    if len(one_hit_wonders) < 2 or len(hitmakers) < 2:
        print("\n  âš ï¸  Insufficient sample size for statistical testing")
        continue
    
    print(f"\n{'-'*90}")
    print(f"{'Metric':<50} {'t-test p':<12} {'MW p':<12} {'Effect Size':<12}")
    print(f"{'-'*90}")
    
    significant_count_t = 0
    significant_count_mw = 0
    
    for metric in metrics_to_compare:
        # Use fillna(0) instead of dropna()
        oww_values = one_hit_wonders[metric].fillna(0)
        hm_values = hitmakers[metric].fillna(0)
        
        if len(oww_values) < 2 or len(hm_values) < 2:
            print(f"{metric:<50} {'N/A':<12} {'N/A':<12} {'N/A':<12}")
            continue
        
        # Independent t-test
        t_stat, t_pval = stats.ttest_ind(oww_values, hm_values, equal_var=False)
        
        # Mann-Whitney U test
        u_stat, mw_pval = stats.mannwhitneyu(oww_values, hm_values, alternative='two-sided')
        
        # Cohen's d effect size
        effect_size = cohens_d(oww_values, hm_values)
        
        # Format p-values
        t_pval_str = f"{t_pval:.4f}" if t_pval >= 0.0001 else "<0.0001"
        mw_pval_str = f"{mw_pval:.4f}" if mw_pval >= 0.0001 else "<0.0001"
        
        # Add significance markers
        t_sig = " *" if t_pval < 0.05 else ""
        mw_sig = " *" if mw_pval < 0.05 else ""
        
        # Effect size interpretation
        if abs(effect_size) < 0.2:
            effect_interp = "negligible"
        elif abs(effect_size) < 0.5:
            effect_interp = "small"
        elif abs(effect_size) < 0.8:
            effect_interp = "medium"
        else:
            effect_interp = "LARGE"
        
        print(f"{metric:<50} {t_pval_str + t_sig:<12} {mw_pval_str + mw_sig:<12} {effect_size:>6.3f} ({effect_interp})")
        
        if t_pval < 0.05:
            significant_count_t += 1
        if mw_pval < 0.05:
            significant_count_mw += 1
    
    print(f"{'-'*90}")
    print(f"Significant results (p < 0.05): {significant_count_t}/{len(metrics_to_compare)} (t-test), {significant_count_mw}/{len(metrics_to_compare)} (Mann-Whitney)")
    print(f"* = p < 0.05 (statistically significant)")
    print(f"Effect size: Positive = Hitmakers higher centrality, Negative = One-Hit Wonders higher")

print("\n" + "="*90)
print("NOTE: This analysis uses fillna(0), treating missing network data as zero")
print("centrality. One-hit wonders are 16-21% less likely to be in collaboration networks.")
print("="*90)
```

```python
# 2. PERCENTILE BREAKDOWN BY 5-YEAR PERIODS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

periods = [
    (1991, 1995, "1991-1995"),
    (1996, 2000, "1996-2000"),
    (2001, 2005, "2001-2005"),
    (2006, 2010, "2006-2010"),
    (2011, 2015, "2011-2015"),
    (2016, 2020, "2016-2020")
]

metric = 'degree_centrality_top10_rolling10'

# Collect results for all periods
all_results = []

for start, end, label in periods:
    df_period = df[(df['first_year_top_10_songs'] >= start) & 
                   (df['first_year_top_10_songs'] <= end)].copy()
    
    df_period[metric] = df_period[metric].fillna(0)
    
    # Define quartiles for non-zero values
    non_zero = df_period[df_period[metric] > 0][metric]
    if len(non_zero) > 0:
        q25 = non_zero.quantile(0.25)
        q50 = non_zero.quantile(0.50)
        q75 = non_zero.quantile(0.75)
    else:
        continue
    
    df_period['centrality_tier'] = pd.cut(df_period[metric], 
                                         bins=[-0.001, 0, q25, q50, q75, 1.0],
                                         labels=['Zero', 'Q1', 'Q2', 'Q3', 'Q4'])
    
    oww = df_period[df_period['top_10_1hit_wonder_song'] == True]
    hm = df_period[df_period['top_10_hitmaker_songs'] == True]
    
    # Calculate percentages in each tier
    oww_dist = oww['centrality_tier'].value_counts(normalize=True).reindex(['Zero', 'Q1', 'Q2', 'Q3', 'Q4'], fill_value=0) * 100
    hm_dist = hm['centrality_tier'].value_counts(normalize=True).reindex(['Zero', 'Q1', 'Q2', 'Q3', 'Q4'], fill_value=0) * 100
    
    all_results.append({
        'period': label,
        'oww_zero': oww_dist['Zero'],
        'oww_q4': oww_dist['Q4'],
        'hm_zero': hm_dist['Zero'],
        'hm_q4': hm_dist['Q4'],
        'oww_dist': oww_dist,
        'hm_dist': hm_dist,
        'oww_n': len(oww),
        'hm_n': len(hm)
    })
    
    # Print for this period
    print("="*80)
    print(f"PERCENTILE BREAKDOWN: {label}")
    print("="*80)
    print(f"\nOne-Hit Wonders (n={len(oww)}):")
    for tier in ['Zero', 'Q1', 'Q2', 'Q3', 'Q4']:
        print(f"  {tier}: {oww_dist[tier]:.1f}%")
    
    print(f"\nHitmakers (n={len(hm)}):")
    for tier in ['Zero', 'Q1', 'Q2', 'Q3', 'Q4']:
        print(f"  {tier}: {hm_dist[tier]:.1f}%")
    
    print(f"\nKEY DIFFERENCES:")
    print(f"  Top Quartile (Q4): {hm_dist['Q4']:.1f}% HM vs {oww_dist['Q4']:.1f}% OHW (Î” = {hm_dist['Q4'] - oww_dist['Q4']:+.1f}pp)")
    print(f"  Zero Centrality:   {oww_dist['Zero']:.1f}% OHW vs {hm_dist['Zero']:.1f}% HM (Î” = {oww_dist['Zero'] - hm_dist['Zero']:+.1f}pp)")
    print()

# Visualize trends across periods
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

periods_labels = [r['period'] for r in all_results]
x = np.arange(len(periods_labels))

# Plot 1: Top Quartile percentages
oww_q4 = [r['oww_q4'] for r in all_results]
hm_q4 = [r['hm_q4'] for r in all_results]

ax1.plot(x, oww_q4, marker='o', linewidth=2.5, markersize=10, color='orange', label='One-Hit Wonders')
ax1.plot(x, hm_q4, marker='s', linewidth=2.5, markersize=10, color='green', label='Hitmakers')
ax1.set_xlabel('Period', fontsize=12, fontweight='bold')
ax1.set_ylabel('% in Top Quartile of Centrality', fontsize=12, fontweight='bold')
ax1.set_title('Top Quartile Centrality Over Time', fontsize=13, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(periods_labels, rotation=45, ha='right')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Plot 2: Zero centrality percentages
oww_zero = [r['oww_zero'] for r in all_results]
hm_zero = [r['hm_zero'] for r in all_results]

ax2.plot(x, oww_zero, marker='o', linewidth=2.5, markersize=10, color='orange', label='One-Hit Wonders')
ax2.plot(x, hm_zero, marker='s', linewidth=2.5, markersize=10, color='green', label='Hitmakers')
ax2.set_xlabel('Period', fontsize=12, fontweight='bold')
ax2.set_ylabel('% with Zero Centrality (No Network)', fontsize=12, fontweight='bold')
ax2.set_title('Non-Networked Artists Over Time', fontsize=13, fontweight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels(periods_labels, rotation=45, ha='right')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

plt.suptitle('Evolution of Centrality Tiers: 5-Year Periods (1991-2020)', 
            fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('percentile_trends_5year_periods.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# MEDIAN COMPARISON: 4 Yearly Centrality Metrics
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('df_artists_top_10_songs_only_with_network_metrics.csv')

periods = [
    (1991, 1995, "1991-1995"),
    (1996, 2000, "1996-2000"),
    (2001, 2005, "2001-2005"),
    (2006, 2010, "2006-2010"),
    (2011, 2015, "2011-2015"),
    (2016, 2020, "2016-2020")
]

metrics = [
    ('degree_centrality_top10_yearly', 'Degree Centrality'),
    ('harmonic_closeness_centrality_top10_yearly', 'Harmonic Closeness'),
    ('betweenness_centrality_top10_yearly', 'Betweenness Centrality'),
    ('eigenvector_centrality_top10_yearly', 'Eigenvector Centrality')
]

# Create 4 subplots (4 rows, 2 columns each: median trends + gap)
fig, axes = plt.subplots(4, 2, figsize=(16, 18))

for metric_idx, (metric, metric_name) in enumerate(metrics):
    # Calculate statistics for each period
    results = []
    
    for start, end, label in periods:
        df_period = df[(df['first_year_top_10_songs'] >= start) & 
                       (df['first_year_top_10_songs'] <= end)].copy()
        
        df_period[metric] = df_period[metric].fillna(0)
        
        oww = df_period[df_period['top_10_1hit_wonder_song'] == True][metric]
        hm = df_period[df_period['top_10_hitmaker_songs'] == True][metric]
        
        results.append({
            'period': label,
            'oww_median': oww.median(),
            'hm_median': hm.median(),
            'difference': hm.median() - oww.median()
        })
    
    results_df = pd.DataFrame(results)
    x = np.arange(len(results_df))
    
    # Left plot: Median trends
    ax1 = axes[metric_idx, 0]
    
    ax1.plot(x, results_df['oww_median'], marker='o', linewidth=2.5, markersize=10, 
            color='orange', label='One-Hit Wonders')
    ax1.plot(x, results_df['hm_median'], marker='s', linewidth=2.5, markersize=10, 
            color='green', label='Hitmakers')
    
    ax1.fill_between(x, results_df['oww_median'], results_df['hm_median'], 
                     where=(results_df['hm_median'] > results_df['oww_median']),
                     alpha=0.2, color='green')
    ax1.fill_between(x, results_df['oww_median'], results_df['hm_median'], 
                     where=(results_df['hm_median'] <= results_df['oww_median']),
                     alpha=0.2, color='orange')
    
    ax1.set_ylabel('Median Value', fontsize=10, fontweight='bold')
    ax1.set_title(f'{metric_name} (Yearly)', fontsize=11, fontweight='bold')
    ax1.set_xticks(x)
    ax1.set_xticklabels(results_df['period'], rotation=45, ha='right', fontsize=9)
    if metric_idx == 0:
        ax1.legend(fontsize=9, loc='best')
    ax1.grid(True, alpha=0.3)
    
    # Right plot: Gap
    ax2 = axes[metric_idx, 1]
    
    bars = ax2.bar(x, results_df['difference'], width=0.6, alpha=0.7, edgecolor='black')
    ax2.axhline(y=0, color='red', linestyle='--', linewidth=2)
    
    # Color bars
    for bar, val in zip(bars, results_df['difference']):
        if val > 0:
            bar.set_color('green')
        else:
            bar.set_color('orange')
    
    # Add value labels
    for i, (xi, val) in enumerate(zip(x, results_df['difference'])):
        ax2.text(xi, val, f'{val:+.4f}', ha='center', 
                va='bottom' if val > 0 else 'top', fontweight='bold', fontsize=8)
    
    ax2.set_ylabel('Gap (HM - OHW)', fontsize=10, fontweight='bold')
    ax2.set_title(f'Difference in Median', fontsize=11, fontweight='bold')
    ax2.set_xticks(x)
    ax2.set_xticklabels(results_df['period'], rotation=45, ha='right', fontsize=9)
    ax2.grid(True, alpha=0.3, axis='y')

# Add overall labels
axes[-1, 0].set_xlabel('Period', fontsize=11, fontweight='bold')
axes[-1, 1].set_xlabel('Period', fontsize=11, fontweight='bold')

plt.suptitle('Median Centrality by Period (1991-2020): 4 Yearly Metrics\nGreen = Hitmakers Higher, Orange = One-Hit Wonders Higher', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('median_comparison_4metrics_yearly_5year.png', dpi=300, bbox_inches='tight')
plt.show()
```
